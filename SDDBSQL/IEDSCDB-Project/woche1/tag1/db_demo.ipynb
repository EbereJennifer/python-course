This code sets up a connection to an SQLite database and prepares something to execute SQL commands.

1. import sqlite3
Loads Python’s built-in SQLite interface
SQLite is a lightweight database stored in a file (no server needed)

2. Create / open the database
connection = sqlite3.connect("verein.db")

This does two things:

If "verein.db" exists → it opens it
If it doesn’t exist → it creates a new database file

👉 So after this line:

You have a file called verein.db
connection represents the link to that database

3. Create a cursor
cursor = connection.cursor()

The cursor is what you use to:

send SQL commands
retrieve results

Think of it like a control tool to talk to the database.

🔹 How it all fits together
connection → manages the database file
cursor → executes SQL queries
🔹 Example of what comes next

Typically you’d do something like:

cursor.execute("CREATE TABLE mitglieder (name TEXT, alter INTEGER)")
connection.commit()
execute() → runs SQL
commit() → saves changes

```python
import sqlite3

connection = sqlite3.connect("verein.db")
cursor = connection.cursor()  # kann SQL-Befehle abschicken

connection.close()
```

```python
import sqlite3

connection = sqlite3.connect("verein.db")
cursor = connection.cursor()

# DDL- statement Data Definition Language
# Erzeugen einer Tabelle

sql_create = """
    CREATE TABLE IF NoT EXISTS mitglieder(
        nummer INTEGER PRIMARY KEY,
        vorname TEXT,
        nachname TEXT,
        geburtstag TEXT,
        betrag REAL
    )
"""
cursor.execute(sql_create)
connection.close()
```

```python
import sqlite3

connection = sqlite3.connect("verein.db")
cursor = connection.cursor()
# DDL-statement Data Definition Language
# Erzeugen einer Tabelle


sql_insert = """
    INSERT INTO mitglieder VALUES(1000, "Bernardo", "Soler", "13.7.1985", 80.50)
"""

cursor.execute(sql_insert)
connection.commit()
connection.close()

```
