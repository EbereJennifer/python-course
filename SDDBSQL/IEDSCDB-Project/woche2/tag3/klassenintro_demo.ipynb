#### Pep8 Standard
According to PEP 8: Top-level definitions (like classes and functions) should be separated by two blank lines.
   - Two blank lines:
        - Between classes
        - Between functions
        - Between class and other top-level code
 - One blank line:
    - Between methods inside a class (optional, often omitted for small classes)



#### class Hund:

This creates a class called Hund (German for “dog”). A class is like a blueprint for creating objects.

 ``__init__ `` is a special method in Python. it is  automatically called when you create a new object from the class.
- Think of it as:

    - “How should a new dog be initialized (set up)?”

    - ``__init__`` = constructor (runs when object is created)
self = the current object
It initializes (sets up) the object’s data

```python
class Hund:
    def __init__(self, rasse, name, farbe):
        self.rasse = rasse
        self.name = name
        self.farbe = farbe


hund1 = Hund("Dackel", "Coco", "braun")  # ruft __int__auf
type(hund1), hund1, hund1.rasse, hund1.name, hund1.farbe
```

```python
class Hund:
    def __init__(self, rasse, name, farbe="unbekannt"):
        self.rasse = rasse
        self.name = name
        self.farbe = farbe


hund1 = Hund("Dackel", "Coco", "braun")  # ruft __int__auf
type(hund1), hund1, hund1.rasse, hund1.name, hund1.farbe
```

```python
hund2 = Hund("Bernhardiner", "Otto") # Anzahl Argumente stimmt
hund2.print_formatted()
hund3 = Hund(name="Mary", rasse="Pudel") # Argumente mit Namen ansprechen
hund3.print_formatted()
hund2, hund3
```

```python
class Hund:
    def __init__(self, rasse, name, farbe="unbekannt"):
        self.rasse = rasse
        self.name = name
        self.farbe = farbe

    def __str__(self):
        return f"Hund({self.rasse}, Rasse: {self.name}, Farbe: {self.farbe})"

    #Methode, nicht notwendig, falls __str__vorhanden ist
    def print_formatted(self):
        print(self)

hund1 = Hund("Dackel", "Coco", "braun")  # ruft __int__auf

type(hund1), hund1, hund1.rasse, hund1.name, hund1.farbe

hund1.print_formatted()
```

```python

#PEP8: Nach klassen 2 zeilen?
# __init__ is called a constructor
class Mitarbeiter:
    # Ü: Konstruktor, Parameter name, jahresbrutto
    def __init__(self, name, jahresbrutto):
        self.name = name
        self.jahresbrutto = jahresbrutto


    def output(self):
        print(f"Mitarbeiter {self.name}, Jahresbrutto: {self.jahresbrutto}")

    def monatsbrutto(self):
        return self.jahresbrutto / 12

    def gehaltserhöhung(self, betrag):
        self.jahresbrutto += betrag


m1 = Mitarbeiter("Guler", 42)
print(m1)
# print(m1.output()) this will give none0
m1.output()
brutto_im_monat = m1.monatsbrutto()
brutto_im_monat
```

One blank line (inside classes)
