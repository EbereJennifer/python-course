```python
"""
Klasse Wettermessung soll Attribute für ort, sonnenaufgang, temperatur enthalten.
10 Objekte dieser Klasse mit Zufallswerten erzeugen und ausgeben (die Daten sollen sichtbar sein).

The WeatherMeasurement class should contain attributes for location, sunrise, and temperature.
Generate and display 10 objects of this class with random values (the data should be visible).

"""
```

```python
import random
from datetime import datetime, timedelta

class wettermessung():
    def __init__(self, location, sunrise, temperature):
        self.location = location
        self.sunrise = sunrise
        self.temperature = temperature


    def output(self):
        s = f"wettermessung {self.location} Sunrise: {self.sunrise} {self.temperature:.2f}C"
        print(s)


locations = ["Hamburg", "Freiburg", "Berlin", "Bonn", "Stuttgart", "Frankfurt", "Dortmund", "Mannheim", "Duesseldorf", "Essen"]
zeitstempel = datetime(2026, 4, 20, 7, 5)
for _ in range(10):
    location = random.choice(locations)
    delta = timedelta(seconds=random.randint(1, 100))
    sunrise = zeitstempel - delta
    temperature = random.uniform(5.0, 15.0)
    w = wettermessung(location, sunrise, temperature)
    w.output()

```
