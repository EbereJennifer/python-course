#### Normal forms (1NF → 6NF)
The “normal forms” are a sequence of rules used in relational database design to reduce redundancy and avoid anomalies (like inconsistent updates or duplicate data). Each higher normal form builds on the previous one.

##### 1st Normal Form (1NF)
Rule:
- No repeating groups or arrays
- Each field contains atomic (indivisible) values
- Each row can be uniquely identified (primary key)
##### 2nd Normal Form (2NF)
Rule:
- Must be in 1NF
- No partial dependency on a composite key
- Every non-key column must depend on the whole primary key, not just part of it. e.g

- Primary key = (student_id, course_id)
- Bad:
- | student_id | course_id | student_name |
- student_name depends only on student_id ❌

    - Fix:
- Move student_name to a separate table
##### 3rd Normal Form (3NF)
Rule:
- Must be in 2NF
- No transitive dependencies
- Non-key attributes must depend only on the key, not on other non-key attributes. e.g
| id | city | postal_code |

- ##### Boyce–Codd Normal Form (BCNF)
Rule:
- Stronger version of 3NF
- Every determinant must be a candidate key
- If something determines another column, it must be a key.

This handles edge cases where 3NF still allows anomalies.

##### 4th Normal Form (4NF)
Rule:
- Must be in BCNF
-No multi-valued dependencies
Example:

| student | hobby | language |

If hobbies and languages are independent: You get unnecessary combinations ❌

Fix:
Split into:
- student–hobby
- student–language

##### ##### 5th Normal Form (5NF)
Rule:
- Must be in 4NF
- No join dependency issues
- Data should not be decomposable into smaller tables unless it can be perfectly reconstructed
- Used in very complex relational designs.

##### 6th Normal Form (6NF)
Rule:
- Tables are reduced to irreducible components
- Often used in temporal databases (tracking changes over time)
- Each table contains:
    - a key
    - a single fact
- Very rare in everyday applications.

##### Practical reality

In real-world systems:

- Most databases aim for 3NF or BCNF
- 4NF+ is mostly for specialized or academic use

##### Quick summary
| Normal Form   |Focus
|---------------|-------------------------------
| 1NF	        |Atomic values
| 2NF	        |No partial dependency
| 3NF	        |No transitive dependency
| BCNF	        |Stronger 3NF
| 4NF	        |No multi-valued dependencies
| 5NF	        |No join dependency issues
| 6NF	        |Fully decomposed (rare)

##### Key idea
Each step removes a specific type of redundancy:

- 1NF → structure
- 2NF → partial duplication
- 3NF → indirect duplication
- higher forms → complex edge cases

##### How to recognize them quickly (exam cheat sheet)
|Symptom	                        |Likely Normal Form Violation
|-----------------------------------|----------------------------
|Multiple values in one column	    |1NF
|Depends on part of composite key	|2NF
|Depends on another non-key column	|3NF
|Determinant is not a key	        |BCNF
|Independent multi-values	        |4NF
|Complex join redundancy	        |5NF
