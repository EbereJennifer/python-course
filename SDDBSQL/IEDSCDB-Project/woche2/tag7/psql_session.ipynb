#### Datenverzeichnis einrichten, Initialisieren
- Beispiel: C:\dev\kurs\sachen_postgres\daten im Windows Explorer erzeugen
-initdb -D "C:\dev\kurs\sachen_postgres\daten" -U sky -E utf8
- pg_ctl -D "C:\dev\kurs\sachen_postgres\daten" -l logfile start
- logfile wird eingerichtet da, wo Eingabeaufforderung gestartet wurde, üblicherweise Homeverzeichnis unter Windows
- Falls nötig, kann der Port für Postgresql in postgresql.conf im Datenverzeichnis geändert werden (Zeile 64, Standard 5432)
- psql -p 56789 -U sky postgres
- psql -p 56789 -U eagbarakwe postgres
