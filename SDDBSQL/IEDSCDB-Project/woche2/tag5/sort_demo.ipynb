```python
# Eingebaute Sortierung durch __lt__
# Bei Aufruf von sort oder sorted: Andere sortierung durch key möglich. Built-in sorting using __lt__
# When calling sort or sorted: Different sorting by key is possible


class Info:
    def __init__ (self, name, jahr):
        self.name = name
        self.jahr = jahr

    def __str__(self):
            return f"{self.name}: {self.jahr}"

    def __lt__(self, other):
            assert isinstance(other, Info)
            return self.name < other.name
```

```python
info1 = Info("Zimmerer", 1965)
info2 = Info("Müller", 2003)
info3 = Info("Gyrim", 1987)
info4 = Info("Schmydt", 1984)
liste = [info1, info2, info3, info4]
for info in liste:
    print(info)
```

```python
# U: Sortierung nach Jahr
for info in sorted(liste, key=lambda elem: elem.jahr):  #key= defines what value is used for sorting
                                                        # lambda elem: elem.jahr means:
                                                        # 👉 “Sort by the jahr attribute of each object”
    print(info)
```

```python
# Python needs comparison methods like:
#
# __lt__(self, other) → “less than”
# __gt__(self, other) → “greater than”

# If your class doesn’t define these, Python doesn’t know how to sort them.

for info in sorted(liste):  #  built-in sorting of the class
    print(info)
```

```python
liste.sort(key=lambda elem: -elem.jahr)
for info in liste:
    print(info)
```

```python
# Funktion zweiter Ordnung mit key als Parameter. # Second-order function with `key` as a parameter
def sort_liste(liste, key):
    liste.sort(key=key)

def print_liste(liste):
    print("*" * 40)
    for info in liste:
        print(info)


sort_liste(liste, lambda elem: elem.jahr)
print_liste(liste)
```

```python
# Ü: sort_liste aufrufen mit einer Liste von Tuple, Sortieren nach Länge. # Call sort_liste with a list of tuples, sorting by length

liste_tupel = [(1,6,4,3), (3,7,8), (3,4), (3,3,3)]

# sort by length:  When two elements have the same key, Python keeps their original order. This is called a stable sort.
liste_tupel.sort(key=len)
print(liste_tupel)

# sort by sum
liste_tupel.sort(key=sum)
print(liste_tupel)
```
