#### Erste Schritte mit Postgres
- Durch das Entpacken ist der Ordner pgsql entstanden
- Bei normaler Installation wird automatisch Pgadmin mitanstalliert, aber nicht mit der Zip-Datei. Mit Pgadmin kann man SQL-Befehle an eine Datenbank schicken, Daten verwalten....
-
- Mitgeliefert wird eine Terminal Anwendung psql. Damit kann man ebenfalls SQL Befehle losschicken
- Anfangs muss eine initialisierung stattfinden.

Innerhalb der Postgres Installation sind die wichtigen Programme im Unterordner bin. Dieser kann zur Umgebungsvariable PATH dazukommen.

- postgres -V
- psql -V
- -Checken in einem Terminal: Windowstaste cmd
