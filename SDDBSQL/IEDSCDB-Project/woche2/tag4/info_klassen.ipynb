### Elemente einer Klasse
#### Instanzbasierte Attribute
- Auch Instanzvariable genannt
- Jedes Objekt der Klasse kann unterschiedliche Werte haben
- Üblicherweise im Konstruktor initialisiert
- Also known as instance variables
- Each object of the class can have different values
- Usually initialized in the constructor
#### Instanzbasierte Methoden
- Instanzmethoden
- Können aud die Instanzvariablen zugreifen: lesen und schreiben
- Brauchen eine Referenz auf eine Instanz (self) und werden im kontext genau dieses einen Objektes ausgeführt
- Can access instance variables: read and write
- Require a reference to an instance (self) and are executed in the context of that specific object
#### Klassenbasierte Attribute
- Auch Klassenvariable genannt
- Gibt es nur einmal pro Klasse
- Lesen und Schreiben ist möglich
#### Klassenbasierte Methoden
- Klassenmethoden
- Sind mit Klassen verbunden, aber nicht mit einem bestimmten Objekt des klasse
- Methode wird mit Dekorator @classmethod markiert, erster Parameter ist hier nicht self, sondern class(cls) für die klasse als Ganzes.
- Are associated with classes, but not with a specific instance of the class
- The method is annotated with the @classmethod decorator; the first parameter here is not self, but class(cls) for the class as a whole
#### Mogische Methoden ((Magic Methods)
- Haben vorgegebene Namen mit __ vorne und hinten
- vordefinierte Bedeutung
- \_\_init\_\_ Konstruktor zum Erzeugen von neuen Objekten. Constructor for creating new objects
- \_\_str\_\_ gibt gut lesbare Stringdarstellung zurück, wird automatisch bei print berücksichtigt. returns a clearly readable string representation, which is automatically taken into account by `print`
- \_\_eq\_\_ soll inhaltiche Gleichheit von Objekten definieren. is intended to define content-based equality between objects
- \_\_del\_\_ Destruktor, wird aufgerufen bei der globalen Funktion del
- \_\_lt\_\_
