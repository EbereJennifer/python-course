##### Autoincrement
When you specify autoincrement, the database automatically generates IDs so you don’t have to supply them.
- Without AUTOINCREMENT, deleted IDs might be reused
- With AUTOINCREMENT, IDs are never reused, always increasing
    - Difference between INTEGER PRIMARY KEY and AUTOINCREMENT
- INTEGER PRIMARY KEY → auto-increments, but may reuse gaps
- INTEGER PRIMARY KEY AUTOINCREMENT → strictly increasing, no reuse

#### Cursor & connection

In your function, cursor is an object that lets your Python code talk to the database.
- ``fetchall()`` pulls everything that’s left in the query result into Python so you can work with it.
- ``cursor.fetchone()`` → gets one row
- ``cursor.fetchmany(size)`` → gets a specific number of rows

#### In CREATE TABLE
cursor.execute(sql_create)
connection.close()
Why no commit() is shown?

Because:

SQLite automatically commits DDL statements (like CREATE TABLE, DROP TABLE, etc.) in many cases
AND closing the connection often triggers an implicit commit

👉 So this works, but it’s slightly “accidental behavior” Even though SQLite may auto-commit, you should not rely on that.

#### In INSERT / Update/ delete
cursor.execute(sql)
connection.commit()
Why commit is required here:
INSERT is DML (data manipulation)
SQLite keeps it in a transaction buffer
If you don’t commit:
the data may NOT be saved to the .db file
or it disappears when the connection closes

👉 So commit() is mandatory for saving inserts reliably

#### __repr__
__repr__ is a special method in Python that defines the official string representation of an object.

It is mainly used for:

- debugging
- development
- showing objects in a precise, unambiguous way
---
in the interactive console (or sometimes in logs), Python tries:
__repr__ (preferred for developers)
__str__ (more user-friendly, if defined)

So:
__repr__ = technical description
__str__ = readable description for users

#### __hash__
__hash__ is a special method in Python that returns an integer “fingerprint” of an object.
This fingerprint is used when you put objects into:

set
dict (as keys)

👉 It allows Python to quickly check if objects are the same.
