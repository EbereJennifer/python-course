### Iteratoren

Objekte, über die man in einer Schleife iterieren kann. Zwei Methoden nötig:

- iter() gibt Iterator zurück
- next()() gibt nächstes Element, wirft am Ende Stoplteration

Ist schon realisiert:

- Listen
- Tupeln

Kann in eigener Klasse realisiert werden.

```python
class ZahlenMitIterator:
    def __init__ (self, top):
        self.current = 1
        self.top = top

    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.top:
            old = self.current
            self.current += 1
            return old
        else:
            raise StopIteration

zahlen = ZahlenMitIterator(top=8)
for n in zahlen:
    print(n)
```

##### Generator
Funktion, die im Inneren mit yield Werte zurückgeben kann. Python macht daraus automatisch einen Iterator.
Sinnvoll mit wenigen bekannten Werten

```python
def mini_generator():
    yield "Python"
    yield "Java"
    yield "Kotlin"
    yield "C#"


gen = mini_generator()
elem1 = next(gen)
elem2 = next(gen)
elem1, elem2
```

```python
gen = mini_generator()
for elem in gen:   # Automatisch: iterator
    print(elem)
```

```python
type(mini_generator), type(mini_generator())
```

```python
# wie wird for Loop hinter den kulissen ausgeführt
liste = ["C++", "Smalltalk", "JSON", "SQL"]
iterator = iter(liste)
while True:
    try:
        value = next(iterator)
        print(value)
    except StopIteration:
        print("Iterieren ist beendet")
        break
print("Weiter gehts...")

```
