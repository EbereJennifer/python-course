```python

filter1 = { "größe": 431 }
filter2 = { "größe": { "$eq": 431 } }
filter3 = { "größe": { "$ne": 431 } }
filter4 = { "größe": { "$lt": 431 } }
filter5 = { "größe": { "$lte": 431 } }
filter6 = { "größe": { "$gt": 431 } }
filter7 = { "größe": { "$gte": 431 } }
filter8 = { "größe": { "$gte": 431, "$lte": 500 } } # beide nötig
filter9 = { "größe": { "$gte": 500 }, "preis": { "$lte": 500_000 } }
filter10 ={ "$or" :[
    {"große": {"gte": 500}},
    {"preis": {"$lte": 500_000}}

    ]
}
filter11 = { "$and": [
    {"große": {"$gte": 500}},
    {"preis": {"$lte": 500_000}}
    ]
}

# Ü: filter10 soll die Bedingungen von filter9 mit ODER verknüpfen

result = coll.find(filter11)
list(result)
```

#### Vergleichsoperatoren
- $eq
- $ne
- $lt
- $lte
- $gt
- $gte
- $in
- $nin
#### Logische Operator
- $and
- $or
- $not
- $nor keine Bedingung darf erfüllt sein
