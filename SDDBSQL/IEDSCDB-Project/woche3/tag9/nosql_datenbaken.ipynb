### NoSQL Datenbanken
Sammelbegriff für unterschiedlichene Techniken
#### Dokument Datenbank
- Collections von Dokumenten innerhalb Datenbank, Schema kann auch innerhalb collection variieren
- Leicht auf Objekte der gängigen Programmiersprachen abbildar
- Am weitesten verbreitet: MongoDB
- CouchDB
#### Key-Value-Store
- Einfachster Typ
- Sehr schnelle Lookups
- Keine komplexen Queries
- Oft Datenspeicherung im RAM
- Genutzt für Caching
- Hauptvertrter: Redis, DynamoDB
-
- #### Column-Store (Wide column)
- Verteilte Systeme
- Sehr große Datenmengen
- Beispiele: Cassandra, HBase
- Nutzen: Big Data, Analytik, Time series data
#### Graph Datenbanken
- Sehr stark verbundene Daten
- Effizient um Beziehung in vielen Schritten durchzugehen
- Beispiele: Neo4j, ArangoDB
- Einsatz: Soziale Netzwerke, Emfehlungssysteme, Betrugserkeenung
