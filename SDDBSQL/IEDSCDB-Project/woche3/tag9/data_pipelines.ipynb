### Data Pipelines
"Verarbeitungsketten": Hoher Grad an Automatisierung, Softwaresysteme, Stichworte: Business Intelligence, Data Lake
Immer beteiligt: Große Datenmangen, die gelesen, transformiert und wieder geschreiben werden.
ETL: Extraction, Transformation, Loading
“Processing chains”: High degree of automation, software systems, keywords: Business Intelligence, Data Lake
Always involved: Large volumes of data that are read, transformed, and rewritten.
ETL: Extraction, Transformation, Loading

##### Datenextraktion
Startschritt, Daten werden aus mehreren Quellen extrahiert:
- Datenbanken
- Dateien, Logs
- APIs, externe Dienst
- First step: Data is extracted from multiple sources:
- Databases
- Files, logs
- APIs, external services

##### Transformation
Oft liegen die Daten noch nicht im optimalen Format vor. Mögliche Shritte:
- Bereingen
- Anwenden von Geschäftslogik
- Aggregieren (Gruppieren, Summenberechnung...)
- Sortieren
- Joinen zwischen datenmengen
Ziel: Geeignetes format für folgenden Schritte
Often, the data is not yet in the optimal format. Possible steps:
- Cleaning
- Applying business logic
- Aggregating (grouping, calculating totals, etc.)
- Sorting
- Joining datasets
Goal: A suitable format for the next steps

##### Datenspeicherung
Verarbeitete Daten werden in Zielsystem gespeichert: Data warehouse
- Gereinigte Daten
- Organisiert
- Strukturiert mit Schemas
Nutzen von data Warehouse:
- Business Intelligenz Dashboards, Übersichtsanzeige
- Business Reporting
- Datenanalyse
- Tools: Tableau, Power BI
- Technologien: redshift (Amazon), BigQuery(Google), Snowflake
- Processed data is stored in the target system: Data warehouse
- Cleaned data
- Organized
- Structured using schemas
Benefits of a data warehouse:
- Business intelligence dashboards, summary views
- Business reporting
- Data analysis
- Tools: Tableau, Power BI
- Technologies: Redshift (Amazon), BigQuery (Google), Snowflake

#### Gegenüberstellung
- Data Warehouse ist für analytische Zwecke geeignet, nicht für OLTP (Live
Transaktionsverarbeitung).
- Im Vergleich zu Data Lake: Warehouse hat strukturierte Daten, optimiert für
Abfragen, Hauptzweck Reporting, BI
- Data Lake sammel daten im Originalformat: Tabellen SQL, CSV, Halbstrukturierte bei JSON, Logdatein, Unstrukturiert Multimedia Audio, video; langsamere Queries, günstiger realisiert z.B.
Amazon S3, Azure Data Lake Storage, Google cloud storage
- A data warehouse is suitable for analytical purposes, not for OLTP (online
transaction processing).
- Compared to a data lake: A data warehouse contains structured data, optimized for
queries, with the primary purpose of reporting and BI
- A data lake collects data in its original format: SQL tables, CSV, semi-structured data (e.g., JSON), log files, unstructured multimedia (audio, video); slower queries, more cost-effective to implement, e.g.,
Amazon S3, Azure Data Lake Storage, Google Cloud Storage
