Ab ca. 2010 breiteten sich neue Arten von Datenbanken aus, genannt NoSQL (Not only SQL), die am weitesten verbreitete ist MongoDB eine dokumentenzentrierte Datenbank (document database).
Vom Python aus spricht man die DB über Biliothen an (Treiber, driver). Zwei grundlegende Varianten:
- Pymongo synchron
- Motor asynchon (hier nicht behandelt).

Pymongo installieren:
pip install pymongo

Starten des Servers:
mongod --dbpath "C:\mein_mongodb\mongo-daten-kurs"
