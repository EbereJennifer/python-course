```python
# ## Task 1
#
# Modify the program so that:
#
# - the name is printed in **uppercase**,
# - if the input is empty, the message **Please enter a name** appears.
#
# ### Note
#
# You must therefore check whether anything has been entered at all.



import tkinter as tk

def begruessen():
    name = entry.get()


    if name.strip() == "":
        label_ausgabe.config(text="Please enter a name")
        return
    name = name.upper()

    label_ausgabe.config(text=f"Hallo, {name}!")


root = tk.Tk()
root.title("Aufgabe1")
root.geometry("400x220")

label_info = tk.Label(root, text="Bitte Namen eingeben:")
label_info.pack(pady=5)

entry = tk.Entry(root)
entry.pack(pady=5)

button = tk.Button(root, text="Begrüßen", command=begruessen)
button.pack(pady=5)

label_ausgabe = tk.Label(root, text="")
label_ausgabe.pack(pady=10)

root.mainloop()
```
