#### UI -> what the user sees,
#### Logic -> what the program knows,

| **UI (View)**                         | **Logic (Model / Domain)**                      | **Glue (Controller)**                           |
| ------------------------------------- | ----------------------------------------------- | ----------------------------------------------- |
| Buttons (Start, Submit, Reset)        | Code checking (is input correct?)               | Button click handlers (`on_submit`, `on_start`) |

| Timer display                         | Timer rules (countdown, limits)                 | Updates UI with results                         |
| Error messages (shown to user)        | Validation rules (e.g. empty input not allowed) | Shows errors in UI                              |
| Layout (window, positioning)          | State (current level, attempts left)            | Controls flow (what happens next)               |
