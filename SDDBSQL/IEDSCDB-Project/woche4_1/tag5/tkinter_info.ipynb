## 1. What is tkinter?

`tkinter` is Python's **standard GUI module**. It allows you to create graphical user interfaces (windows, input fields, buttons, etc.) without having to install additional third-party libraries.

### What is tkinter used for?
With `tkinter`, you can, for example:

- Build input forms,
- Create small desktop tools,
- Develop tutorials or prototypes,
- Control data via buttons and input fields,
- Learn the basics of event-driven programming.

### Why is tkinter suitable for beginners?
- It usually comes bundled with Python.
- It is straightforward.
- Many basic GUI concepts can be easily learned with it:
  - Windows
  - Widgets
  - Layout
  - Events
  - Data binding

### Basic structure of a tkinter program
Almost every tkinter program includes these three steps:

1. Create a window
2. Create widgets
3. Start the event loop (`mainloop()`)

> Important: From an educational standpoint, tkinter is excellent for beginners. For large, modern production applications, there are often other GUI frameworks. However, tkinter is just right for starting a course.

## Mini Mental Model: What Happens in Tkinter?

A Tkinter app usually consists of four parts:

1. **Import**: We import Tkinter into the program.
2. **Main Window**: We create a window using `tk.Tk()`.
3. **Widgets**: We place elements in the window, such as labels, buttons, and input fields.
4. **Event loop**: With `mainloop()`, the program waits for clicks, keyboard input, etc.

### Whiteboard diagram

```text
Python program
     |
     v
Window root = tk.Tk()
     |
     +--> Label
     +--> Entry
     +--> Button --Click--> Function
     |
     v
root.mainloop()  =  wait for events
```

Important: `mainloop()` is not just for show. Without this loop, the window will not respond continuously.
But without mainloop(), Tkinter never starts its event loop.
The script reaches the end and terminates immediately.

So you may see:

no window at all,
or a window that flashes briefly and disappears.

mainloop() is the part that:

keeps the window open,
listens for events (clicks, keyboard input, resizing),
updates the GUI continuously.


#### 4. First Widgets: Label, Button, Entry

Widgets are the visible building blocks of a GUI.

#### Three Important Basic Widgets
- `Label`: displays text
- `Button`: triggers an action
- `Entry`: text input field

- The `pack()` function actually places the label in the window.
- Important:
- A widget only becomes visible once it is placed in the window using a layout manager
- such as `pack()`, `grid()`, or `place()`.
#
- `pady=10` means:
- A margin of 10 pixels is added at the top and bottom.
label.pack(pady=10)

- An Entry is a single-line text field.
- Users can type text into it.
#
- Here, too, “root” means:
- The text field belongs to the main window.
entry = tk.Entry(root)


- The text field must also be positioned in the window using pack().
entry.pack(pady=10)


- A Button is a button.
- For now, this button only has text.
#
- Also important:
- This button doesn’t do anything yet,
- because we haven’t assigned it a function using command=...
button = tk.Button(root, text="Click me")


- The button is also placed in the window using pack().
button.pack(pady=10)
