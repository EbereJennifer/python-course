```python
import tkinter as tk
from tkinter import TkVersion

from dns import name

print("tkinter importiert.")
print("version des Tk-Systems:", TkVersion)
```

```python
root = tk.Tk()                                  # To create a main window using `tk.Tk()`The name root is a convention.
                                                    # You could also call it window.
root.title("Mein erstes Tkinter-Fenster")
root.geometry("400x200")                                    # Format: “width x height”, e.g., 400 pixels wide and 200 pixels high.
label = tk.Label(root, text="Willkommen zu tkinter!")       # the label of the window is placed using pack()
label.pack(pady=10)
entry = tk.Entry(root)
entry.pack(pady=10)                                             # `pady=10` adds vertical spacing.
button = tk.Button(root, text="Klick mich")
button.pack(pady=10)                                    # place button in window using pack()


root.mainloop()                                         # Start the event loop: From here on, tkinter waits for events e.g mouse clicks,                                                        # keyboard input, window close, etc.
```

```python
"""
Wir planen eine kleine Begrüßungs-App, aber noch ohne Code.

Die App soll:

- einen Namen eingeben lassen,
- einen Button zum Begrüßen haben,
- eine Ausgabe anzeigen.


"""

def begruessen():                               # The function will only run when the button is clicked.'
    name = entry.get()               #entry is the text box.and .get() takes whatever the user typed and stores it in the variable name.
    if name.strip() == "":              # Check if the input is empty, .strip() removes spaces before and after the text.
        label_ausgabe.config(text="Bitte gib zuerst einen Namen ein.")
        return
    label_ausgabe.config(text=f"Hallo, {name}!")     # .config() updates widget settings.


root = tk.Tk()
root.title("Begrüßungs-App")
root.geometry("400x220")

label_info = tk.Label(root, text="Bitte Namen eingeben:")  # Create a label. A label is a piece of text displayed in the window.
                                    #root = place it inside the main window.
                                    # text= = text to display.

label_info.pack(pady=5)        #.pack() adds the widget to the window. and # pady=5 adds vertical spacing (5 pixels above and below).

entry = tk.Entry(root)   # This creates the box where the user can type a name (input field)
entry.pack(pady=5)

button = tk.Button(root, text="Begrüßen", command=begruessen)
button.pack(pady=5)


label_ausgabe = tk.Label(root, text="")  # Create an empty output label
label_ausgabe.pack(pady=10)       # Display the output label, Adds it to the window.

root.mainloop()   # Start the event loop Without this line, the window would close immediately.
```
