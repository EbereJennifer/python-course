```python
class verwaltetergeldbetrag:
    def __init__(self, betrag=0.0):
        self.betrag = betrag
    def einzahlen(self, delta):
        b_am = self.einzahlen_moeglich(delta)
        if b_am:
            self.betrag += delta
        else:
            print("Einzahlung nicht moeglich")
        return b_am

    def auszahlen(self, delta):
        b_am = self.einzahlen_moeglich(delta)
        if b_am:
            self.betrag -= delta
        else:
            print("Auszahlung nicht moeglich")
        return b_am

    def einzahlen_moeglich(self, delta):
        if delta <= 0:
            return False
        else:
            return True
    def auszahlen_moeglich(self, delta):
        if delta <= self.betrag:
            return True
        else:
            return False

class Allgemeinekonto(verwaltetergeldbetrag):
    def __init__(self, inhaber, kontonummer, kontostand=0.0):
        super().__init__(kontostand)
        self.inhaber = inhaber
        self.kontonummer = kontonummer

    def geldtranfer(self, ziel, delta):
        if self.auszahlen_moeglich(delta) and ziel.einzahlen_moeglich(delta):
            ziel.einzahlen(delta)
            self.auszahlen(delta)
        else:
            print("Ueberweisung nicht moeglich")

class AllgemeinesKontoMitTagesumsatz(Allgemeinekonto):
    def __init__(self, inhaber, kontonummer, kontostand=0.0, max_tagesumsatz=1000.0):
        super().__init__(inhaber, kontonummer, kontostand)
        self.umsatz_heute = 0.0
        self.max_tagesumsatz = max_tagesumsatz

    def auszahlen(self, delta):
        #if self.umsatz_heute + delta > self.tagesumsatz:
        super().auszahlen(delta)
        self.umsatz_heute += delta

    def naechster_tag(self):
        self.umsatz_heute = 0.0


```
