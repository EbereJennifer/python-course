```python
class Konto:
    def __init__(self, kontoinhaber, kontonummer, zinssatz, startguthaben=0):     # starting balance
        self.kontoinhaber = kontoinhaber
        self.kontonummer = kontonummer
        self.zinssatz = zinssatz
        self.kontostand = startguthaben

    def __str__(self):
        return (f"Konto von {self.kontoinhaber} "
                f"Nr: {self.kontonummer} "
                f"Kontostand: {self.kontostand:.2f}€ "
                f"Zinssatz: {self.zinssatz}%")


    def einzahlen(self, betrag):                    # deposit
        if betrag <= 0:
            raise ValueError("Einzahlungsbetrag muss positiv sein")
        self.kontostand += betrag


    def abheben(self, betrag):                          #withdraw
        if betrag <= 0:
            raise ValueError("Abhebungsbetrag muss positiv sein")

        if self.kontostand - betrag < 0:
            raise ValueError("Nicht genug Guthaben – Konto darf nicht negativ werden!")

        self.kontostand -= betrag

    # Zinsen berechnen und hinzufügen
    def zinsen_gutschreiben(self):
        zinsen = self.kontostand * (self.zinssatz / 100)
        self.kontostand += zinsen

konto = Konto("Max Müller", "FE123456", 2.5, 100)
print(konto)
konto.einzahlen(50)
print(konto)
konto.abheben(120)
print(konto)
konto.zinsen_gutschreiben()
print(konto)
#konto.abheben(50)
#print(konto)
```
