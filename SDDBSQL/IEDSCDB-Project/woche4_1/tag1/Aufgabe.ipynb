## Aufgabe 1.1 – Klasse `Buch`

Implementiert eine Klasse `Buch` mit mindestens diesen Attributen:
- `titel`
- `autor`
- `ausgeliehen_von` (am Anfang `None`)

Implementiert mindestens diese Methoden:
- `ausleihen(name)`
  Leiht das Buch an eine Person aus.
- `zurueckgeben()`
  Gibt das Buch zurück.
- `info()`
  Gibt einen gut lesbaren Text zurück.

### Regeln
- Ein bereits ausgeliehenes Buch darf **nicht nochmal** ausgeliehen werden.
- Ein Buch, das nicht ausgeliehen ist, kann nicht sinnvoll zurückgegeben werden.
- Ihr könnt mit `True` / `False` arbeiten oder Fehlermeldungen ausgeben.

### Mini-Testideen
- Buch erzeugen
- ausleihen
- doppelte Ausleihe verhindern
- zurückgeben
"""


```python
class Buch:

    def __init__(self, titel, autor):
        self.titel = titel
        self.autor = autor
        self.ausgeliehen_von = None

    def __str__(self):
        return f"Titel: {self.titel}, Autor: {self.autor}, ausgeliehen_von{self.ausgeliehen_von}"

    def ausleihen(self, name):
        if self.ausgeliehen_von is not None:
            return f"Das Buch wurde bereits ausgeliehen von {self.ausgeliehen_von}"

        self.ausgeliehen_von = name
        return f"Das Buch wurde von {name} ausgeliehen"

    def returned_buch(self):
        if self.ausgeliehen_von is None:
            return f"Das Buch ist erhältlich"
        name = self.ausgeliehen_von
        self.ausgeliehen_von = None
        return f"{name} hat das Buch zurückgegeben."

    def info(self):
        if self.ausgeliehen_von is None:
            status = "verfügbar"
        else:
            status = f"ausgeliehen von {self.ausgeliehen_von}"

        return f"Titel: {self.titel}, Autor: {self.autor}, Status: {status}"



buch = Buch("The Golden Age", "Harry Potter")
print(buch.info())


print(buch.ausleihen("Anna"))
# Das Buch wurde von Anna ausgeliehen.

print(buch.ausleihen("Tom"))                    # the book is already borrowed (by Anna)
# Das Buch ist bereits ausgeliehen.

print(buch.info())

print(buch.returned_buch())

print(buch.info())
```
