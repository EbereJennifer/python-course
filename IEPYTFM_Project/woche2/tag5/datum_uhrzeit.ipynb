#### isoformat()
converts a date or datetime object into a string using the ISO 8601 standard format. YYYY-MM-DD
- .isoformat() always returns a string
- You cannot change its format (unlike strftime())

```python
# weltweit verstandenes standardformat, als string automatisch richtig sortiert
datum.isoformat()
```

```python
from datetime import date

einige_datumwerte = [date(2026, 6, 28), date(2024, 12, 5), date(2024, 12, 3)]
sorted(einige_datumwerte)
```

```python
d1 = date(2026, 3, 17)
d2 = date(2026, 4, 29)
diff = d2 - d1                   # How many days are between these two dates?”
type(diff), diff                # So diff is a timedelta object type(diff)
                                # A timedelta can store:
                                    # days
                                    # seconds
                                    # microseconds
```

```python
diff.seconds, diff.total_seconds()
```

What you can do with timedelta
Add time to a date
from datetime import timedelta

d1 + timedelta(days=10)

- Gives a date 10 days later

Subtract time
d2 - timedelta(days=5)

- 5 days earlier

##### Summary
d2 - d1 → gives a timedelta
A timedelta = time difference
diff.days → number of days
Useful for:
calculating durations
deadlines
age, intervals, etc.

#### strftime
string format time converts a date/time object into a formatted string. You choose how the date should look as text
| Code | Meaning        | Example   |
| ---- | -------------- | --------- |
| `%d` | Day (2 digits) | 08        |
| `%m` | Month          | 04        |
| `%Y` | Year (full)    | 2026      |
| `%y` | Year (short)   | 26        |
| `%A` | Weekday name   | Wednesday |
| `%B` | Month name     | April     |
|`"%b"`| short form     | Apr       |
| `%j` |April 8         |is the 98th day of the year- Always 3 digits (leading zeros!)
|`"%p"`| shows time     | AM / PM   |
"""



```python


today.strftime("%d.%m.%Y")    # string-format-time, deutches Format
```

```python
# Ü: Iso Datumsformat mit strftime nachbauen

today.strftime("%Y-%m-%d-")
```

##### .strptime(): String Parse Time
is basically the reverse of .strftime(). It converts a string into a datetime object according to a specified format.
- e.g `from datetime import datetime
        datetime.strptime(date_string, format)`

- date_string → the date/time as text
- format → how Python should interpret that text (like %Y-%m-%d)

```python
#Americanisches format

from datetime import datetime

erg = datetime.strptime("08/30/2007 14:27:53", format)
erg
```
