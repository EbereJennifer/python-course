```python
# Anton
"""
Datumsobjekt erster_tag für ersten Tag unseres Kurses (letzten Dienstag) erzeugen,
in drei verschiedenen Formaten ausgeben z.B. Iso, deutsches und US-Format.
Zu erster_tag mehrere Tage (z.B. 2, 5 und 8) als Delta
dazuzählen und abziehen und jeweils Ergebnisdatum einschließlich Wochentag anzeigen.
"""

from datetime import date, timedelta # kurze Namen erlaubt

erster_tag = date(2026, 3, 31)
print(erster_tag.isoformat())

format_de = "%d.%m.%Y"
format_us = "%m.%d.%Y"

print(erster_tag.strftime(format_de))
print(erster_tag.strftime(format_us))
```

```python
format_de_mit_wochentag = "%A, %d.%m.%Y"
```

```python
for n in [2, 5, 11]:
    # jeweils delta erzeugen für n Tage, delta zu datum dazuzählen und abziehen
    delta = timedelta(days=n)
    danach = erster_tag + delta
    davor = erster_tag - delta
    print(danach.strftime(format_de_mit_wochentag))
    print(davor.strftime(format_de_mit_wochentag))


```

```python

#Tag und Uhrzeit nacheinander als Input im deutschen Format einlesen,
#ein datetime Objekt daraus machen.

import datetime # lange Namen

string_datum = input("Eingabe Datum: ")
string_uhrzeit = input("Eingabe Uhrzeit: ")

string = f"{string_datum} {string_uhrzeit}"

format_de = "%d.%m.%Y %H:%M"
dt = datetime.datetime.strptime(string, format_de)


print(dt.strftime(format_de))
```

```python
"""
4)
Mit einem datetime Objekt mehrere Rechenoperationen durchführen und Ergebnis anzeigen (z.B. eine Stunde dazuzählen, einige Minuten abziehen usw).
Eine Liste mit Zeitpunkten mit der map-Funktion umwandeln, so dass alle Elemente 10 Minuten später sind.

4)
Perform multiple arithmetic operations with a datetime object and display the result (e.g., add one hour, subtract a few minutes, etc.).
Convert a list of timestamps using the map function so that all elements are 10 minutes later.

"""


from datetime import datetime, timedelta

zeitpunkt = datetime(2024, 10, 13, 17, 25)
abstand1 = timedelta(hours=2, minutes=10)
abstand2 = timedelta(minutes=3, seconds=28)

zeitpunkt_zukunft = zeitpunkt + abstand1
zeitpunkt_vergangenheit = zeitpunkt - abstand2


print(zeitpunkt)
print(zeitpunkt_zukunft)
print(zeitpunkt_vergangenheit)

liste = [zeitpunkt, zeitpunkt_zukunft, zeitpunkt_vergangenheit]
print(liste)

abstand_10_min = timedelta(minutes=10)

print()

map_erg = map(lambda dt: dt + abstand_10_min, liste)
for elem in map_erg:
    print(elem)
```
