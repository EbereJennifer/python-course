#### Fehlerbehandlung
Man will vermeiden, dass ein Programm unnötigerweise abstürzt. Einige Arten von Fehlern können behandelt
werden. Grundkonstrukt: try - except z.B. falls eine Zahl erwartet wird und etwas anderes eingegeben wird.
One wants to avoid a program crashing unnecessarily. Some types of errors can be handled. Basic construct: try - except, e.g., if a number is expected and something else is entered.

- What is the relevance of try and except? In many languages (especially Python), try / except is used to handle errors without crashing your program. Think of it like this:
- `try` → “Attempt to run this code”
- `except` → “If something goes wrong, handle it like this instead”
-  You mustn't always use `try` and `except` in your code and using it everywhere is actually bad practice.
- Use it when: Something is likely to fail (user input, files, APIs, etc.) and you can recover from the error

##### Don't Use it when:
- You’re hiding bugs
- You expect the code to always work
- You’re using it instead of fixing logic errors

#### How do you know which except to use?
Different errors have different types. You choose based on what might go wrong.

- Common ones:
    - ValueError: When the value is the right type but wrong format
    -`int("abc")`  # ❌ ValueError
    - NameError: When a variable is not defined
    - `print(x)`  # ❌ NameError
    - ZeroDivisionError
    - 10 / 0  # ❌
    - KeyboardInterrupt: When the user presses Ctrl + C

#### Bonus tip
- You can also combine errors:

- `except (ValueError, ZeroDivisionError):`
        `print("Invalid input")`
- Quick summary:
    - try/except = prevent crashes + handle errors
    - Don’t overuse it
    - Catch specific exceptions, not everything
    - Use it mainly for unpredictable situations



```python
eingabe = input("Eingabe der Jahreszahl:")
try:
    jahr = int(eingabe)
except ValueError: # möglichst Errortyp angeben
    print("Keine gültige Jahreszahl:", eingabe)
```

Mehrere except Blöcke: Multiple except blocks

```python
try:
    zahl1 = float(input("Zahl1:"))
    zahl2 = float(input("Zahl2:"))
    zahl3 = zahl1 / zahl2
    print(zahl1, zahl2, zahl3)
except ValueError:
    print("Ungültige Eingabe für Zahl")
except ZeroDivisionError:
    print("Division durch 0 ist nicht erlaubt")


print("Nachher")
```

```python
try:
    zahl1 = float(input("Zahl 1"))
    zahl2 = float(input("Zahl 2"))
    zahl3 = zahl1 / zahl2
    print(zahl1, zahl2, zahl3)
except ValueError as error: #Name error ist frei wählbar, auch gebräuchlich e, ex: Name error can be freely chosen, also common e, ex
    print(str(error)) # oder: print(error)
except ZeroDivisionError as error:
    print(str(error))

print ("Nachher")

```

```python
# Ü: Funktion read_int_non_zero soll Input einlesen, bis eine gültige Zahl != 0 vorliegt. is present

def read_int_non_zero():
    while True:
        try:
            zahl = int(input("enter a number greater than zero: "))
            if zahl != 0:
                return zahl
            else:
                print("zero integer or value is not allowed here.")
                continue
        except ValueError:
            print("Invalid input. Try entering the right integer value again.")



n = read_int_non_zero()
n

```

#### class ... (Exception)
This means: “Create a new type of error based on the built-in Exception.”
- So instead of using generic errors like: ValueError, TypeError
- You define your own:
    - InvalidMailAddress
- Because it makes your code more meaningful and precise.

##### Big picture
- class InvalidMailAddress(Exception)
    - Create your own error type
- pass
    - No extra logic needed (just inherit behavior)
- raise
    - Trigger the error
- except InvalidMailAddress
    - Catch only that specific error
- finally
    - Always runs
5) What is finally?

finally means:

“Run this code no matter what happens.”

- ✔ Error or no error
- ✔ Return or no return
- ✔ Crash or success

It’s often used for:

- closing files
- cleaning up resources
- logging

```python
# Eigene Fehlerklasse # Custom error class
class InvalidMailAddress(Exception):
    pass       # pass literally means: “Do nothing.”, it means “Create the class, but don’t add any extra behavior.”

def check_mail_data(username, mail, preferences):
    if not "@" in mail:
        raise InvalidMailAddress("mail muss @ enthalten: " + mail) # this works perfectly, because your class
                                                                    # already inherits everything from Exception.
try:
    check_mail_data("anton", "antonqweb.de", None)
except InvalidMailAddress as ex:                         # ex contains the message
    print(ex)
finally:
    print("Immer ausgeführt: bei finally")
```

```python
# Weiteres Beispiel für eigene Exceptionklasse
# Modernes Python: Camel case für KlassenNamen

from datetime import date, timedelta
class NumberOfDaysNegative(Exception):
    pass


def get_later_date(origin, days):
    if days < 0:
        raise NumberOfDaysNegative()
    return origin + timedelta(days=days)   # remember this replaces having to use else, by placing return in the space for else


get_later_date(date.today(), 5)
# get_later_date(date.today(), -5) löst Exception aus (triggers an exception)
```
