The math library or module
offers a rich selection of mathematical functions and constants.
- Rounding up, rounding down
- Power, square root
- Trigonometric functions sine, cosine, tangent
- Exponential function and logarithm
- GCD, gcd (greatest common divisor), factorial


```python
# Ceil = “ceiling” → round UP
# It returns the smallest integer greater than or equal to a number.

# Floor = “ground” → round DOWN
# It returns the largest integer less than or equal to a number.
import math

math.floor(4.7), math.floor(-2.3), math.ceil(4.2), math.ceil(-2.3),

```

```python
math.pow(2.5, 7.1)   # (2.5 ** 7.1)
math.pow(2.5, 7.1), math.sqrt(7)
```

```python
# Winkel in Grad: 90 Grad bedeutet rechter Winkel(angle)
# Winkel in Radius: 2 * oi vollen kreis

winkel = math.radians(90)
math.sin(winkel), math.cos(winkel), math.tan(winkel)
```

```python
winkel = math.radians(89)
math.sin(winkel), math.cos(winkel), math.tan(winkel)
```

```python
# exponential funktion

math.exp(3) # e hoch 3
```

```python
math.log(3), math.log(100), math.log10(50), math.log2(1000)
```
