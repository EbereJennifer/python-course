```python
# Fehlt: klarmachen, dass Argument >= 0 erwartet wird

def schlechte_function(n):
    if n == 0:
        print("Ende erreicht")
    else:
        # function str wandelt Argument in string um
        print("Wir sind jetzt bei " + str(n))
        schlechte_function(n - 1)


schlechte_function(3)
```

#### Filter
filter(function, iterable): filter, Goes through each element in the list and Keeps only the elements where the function returns True
- What None means here: When you pass None as the function,
- filter(None, iterable)
- It automatically removes all “falsy” values: Falsy values in Python include:
    - 0
    - None
    - False
    - "" (empty string)
    - [], {} (empty collections)

```python
result_list = list(filter(None, [1, 6, 8, 3, 7]))
result_list
```

```python
help(filter)
```

```python
# filter: Erstes Argument ist Function, die Wahrheitswert liefert
result_list = list(filter(lambda elem: elem > 5, [1, 6, 8, 3, 7]))
result_list
```

#### Map
map is quite similar, but all types are allowed as a result in Lambda

```python
# map ist ganz ähnlich, aber alle Typen als Ergebnis bei Lambda erlaubt
# Die Anzahl der Elemente ändert sich hier nicht
result_list = list(map(lambda elem: elem > 5, [1, 6, 8, 3, 7]))
result_list
```

```python
# filter removes 0
result_list = list(filter(lambda elem: elem, [1, 6, 0, -1, -2, 0, 8, 3, 7]))
result_list
```
