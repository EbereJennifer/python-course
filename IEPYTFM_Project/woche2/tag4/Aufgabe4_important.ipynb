#### What the function in Aufgabe 4 is trying to do
It splits a total number (anzahl_buchstaben) into 4 categories:

- lowercase letters (anzahl_low)
- uppercase letters (anzahl_upp)
- punctuation (anzahl_punct)
- digits (anzahl_dig)
🧩 Step-by-step
##### Function input
give_me_a_random_username(10)
- anzahl_buchstaben = 10

##### Initialize counters
- anzahl_low = 0
- anzahl_upp = 0
- anzahl_punct = 0
- anzahl_dig = 0

 These variables will store how many characters of each type you want.

##### Divide into 4 equal parts
- standard = anzahl_buchstaben // 4
- // means integer division (no decimals): 10 // 4 = 2
- Each category gets at least 2 characters

##### Assign equal parts
- anzahl_upp = standard
- anzahl_punct = standard
- anzahl_dig = standard

##### Now
- uppercase = 2
- punctuation = 2
- digits = 2

##### Give the remainder to lowercase
- anzahl_low = anzahl_buchstaben - (3 * standard)

##### Why?

- Because you already assigned: 3 * 2 = 6 characters

- So: anzahl_low = 10 - 6 = 4

##### Final result
print(anzahl_low, anzahl_upp, anzahl_punct, anzahl_dig)

- Output: 4 2 2 2
Why this logic? Because 10 is not divisible by 4: Equal split would be 2.5 each ❌ not possible. So:
- Give 2 to each group
- Put the leftover (4) into lowercase

##### Important note
The function does NOT actually generate a username yet. Even though you imported random, it:

- doesn’t create characters
- doesn’t build a string
- It only calculates how many of each type you should have
What it’s probably meant for

##### Next step would be something like:

1. generate random lowercase letters
2. generate uppercase letters
3. generate digits
4. generate punctuation
5. combine them into a username

###### pseudo idea
- lowercase_letters = ...
- uppercase_letters = ...
- digits = ...
- symbols = ...

```python
# Anton
# Aufgabe 4

import random

def give_me_a_random_username(anzahl_buchstaben):
    anzahl_low = 0
    anzahl_upp = 0
    anzahl_punct = 0
    anzahl_dig = 0

    standard = anzahl_buchstaben // 4
    anzahl_upp = standard
    anzahl_punct = standard
    anzahl_dig = standard
    anzahl_low = anzahl_buchstaben - (3 * standard)
    print(anzahl_low, anzahl_upp, anzahl_punct, anzahl_dig)


    erg = []
    for _ in range(anzahl_upp):
        upper = random.choice(string.ascii_uppercase)
        erg.append(upper)

    for _ in range(anzahl_upp):
        lower = random.choice(string.ascii_uppercase)
        erg.append(upper)

    for _ in range(anzahl_low):
        lower = random.choice(string.ascii_lowercase)
        erg.append(lower)

    for _ in range(anzahl_punct):
        punct = random. choice(string.punctuation)
        erg.append(punct)

    for _ in range(anzahl_dig):
        digit = random.choice(string.digits)
        erg.append(digit)

    random.shuffle(erg)
    #return erg
    return "".join(erg)  #this works


give_me_a_random_username(5)

give_me_a_random_username(10)
```

The function trig_tabelle is generic:

def trig_tabelle(trig_fun):

👉 You can pass any trigonometric function:

trig_tabelle(math.sin)
trig_tabelle(math.cos)
trig_tabelle(math.tan)

And the output will automatically adjust:

sin an der Stelle ...
cos an der Stelle ...
tan an der Stelle ...

```python
def trig_tabelle(trig_fun):
    for grad in range(0, 190, 10):
        rad = math.radians(grad)
        y = trig_fun(rad)
        print(f"{trig_fun.__name__} an der Stelle {grad:3} Grad: {y:8.6f}")

# __name__ gives you the name of a function as a string.
# e.g trig_tabelle(math.sin) : trig_fun.__name__ becomes sin i.e trig_tabelle(math.sin)

print("*" * 40)
trig_tabelle(math.cos)
```
