```python

# Ü: 100 mal Würfel simulieren, Ergebnisse in ein Dictionary,
# als Value die Häufigkeit eintragen z.B. 2: 17, ..., 5: 13

# Simulate rolling a die 100 times, store the results in a dictionary,
# and enter the frequency as the value, e.g., 2: 17, ..., 5: 13

import random

results = {}

for _ in range(100):
    roll = random.randint(1, 6)

    if roll in results:
        results[roll] += 1
    else:
        results[roll] = 1
print(results)

```

```python
# Chat gpt
import random

results = {i: 0 for i in range(1, 7)}

for _ in range(100):
    roll = random.randint(1, 6)
    results[roll] += 1

print(results)
```

#### Kommazahlen
Funktion random() liefert einen Wert zwischen 0.0 und 1.0, bei uniform() sind alle Werte gleich wahrscheinlich, bei gauss() ergibt sich eine Normalverteilung (Glockenform).

```python
# Anton

random.random(), random.uniform(0.5, 4.5), random.gauss(3.0, 1.0)
```

#### Auswahl aus einer Sequenz
- choice() liefert einzelnes Elemnt
- sample() liefert mehrere Elemente ( <= Gesamtzahl)
- choices() beliebig viele Elemente, evt.mit Wiederholungen

```python

pizzas = ["Margherita", "Calzone", "Futti de mare", "Capricchios"]
random.choice(pizzas), random.sample(pizzas, 2), random.sample(pizzas, 4)
```

```python
random.choices(pizzas, k=5)
```

```python
pizzas = ["Margherita", "Calzone", "Futti de mare", "Capricchios"]
random.choices(pizzas, weights=[2, 3, 1, 2], k=10)
# What this means
# Calzone will appear most often
# Futti de mare will appear least
# The others are in between
```
