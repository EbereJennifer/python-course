```python
# Beispiel: liste erst filter, dann transformieren
# 1. Schritte: Texte ausfiltern, die "Python" (klein/Großschreibung)enthalten
# 2. Schritt: Verbleibende Texte ganz in kleinbuchstaben


#Lösung mit filter, map
startliste = ["Python ist toll", "Java", "pYThon? wirklich", "kotlin rules"]
```

```python
filter_result = filter(None, startliste)
gefilterte_liste = list(filter_result)
gefilterte_liste
```

```python
filter_result = filter(lambda text: "python" in text.lower(), startliste)
gefilterte_liste = list(filter_result)
gefilterte_liste
```

```python
map_result = map(lambda text: text.lower(), gefilterte_liste)
transformierte_liste = list(map_result)
transformierte_liste
```

```python
# Zweite Lösung mit list comprehension
[text for text in startliste]
```

```python
[text.lower() for text in startliste if "python" in text.lower()]
```
