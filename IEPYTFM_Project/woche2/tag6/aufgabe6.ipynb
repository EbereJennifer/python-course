```python
# 1)Listcomprehension: Ausgehend vom Intervall 1-100 eine neue Liste mit den Quadratwurzeln aller Primzahlen erzeugen.
# 1)List comprehension: Starting from the interval 1-100, create a new list with the square roots of all prime numbers.
```

```python
import random

from math import sqrt

random.seed(2)
list = random.sample(range(1,101), 100)

list_comprehension = [sqrt(elem) for elem in list if elem % 2 != 0]
list_comprehension_sort = sorted(list_comprehension)

print(list_comprehension_sort, end=" ")


```

```python
"""
4)
Funktion create_random_dates_file(start_date, end_date, count, file_name)
soll zufällige Datumswerte (Typ datetime.date) im Bereich von start_date bis end_date erzeugen und zeilenweise in einer Textdatei speichern.
Funktion read_dates(file_name) soll aus der Datei wieder eine Liste mit Datumswerten machen.
"""
from datetime import datetime, timedelta
from random import randint

today = date.today()

def random_date(start_date, end_date):
    delta = end_date - start_date
    top = delta.days
    zufall = randint(0, top)   #top inclusive
    datum = start_date + timedelta(days=zufall)
    return datum


for _ in range(5):
    d = random_date(today, today + timedelta(days=365))
    print(d)

```

```python
from datetime import datetime, timedelta
from random import randint

today = date.today()
later = today + timedelta(days=200)

def random_date(start_date, end_date):
    delta = end_date - start_date
    top = delta.days
    zufall = randint(0, top)   #top inclusive
    datum = start_date + timedelta(days=zufall)
    return datum

def create_random_dates_file(start_date, end_date, count, file_name):
    with open(file_name, "w") as file:
        for i in range(count):
            datum = random_date(start_date, end_date)
            file.write(date.isoformat(datum) + "\n")


create_random_dates_file(today, later, 15, "random_dates.txt")



def read_dates(file_name):
    with open(file_name, "r") as file:
        return [date.fromisoformat(line.strip()) for line in file]


erg = read_dates("random_dates.txt")
print(erg)


```
