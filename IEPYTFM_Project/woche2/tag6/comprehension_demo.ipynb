#### Comprehensions
- This is a special synthax in Python for creating Lists, sets or Dictionaries
- zusätzlich können Elemente durch einen Filter ausgefiltert werden (if-bedingung am Ende)
-

```python
# List Comprehension erzeugt oft neue Liste aus einer alten

alt = [1, 2, -45, 123, 65, -99, 0, 13]
neu1 = [elem for elem in alt]
neu2 = [elem > 0 for elem in alt]
neu3 = [elem * 5 - 3 for elem in alt]   # the symbol * denotes the element that you have to start with


alt, neu1, neu2, neu3
```

```python
lower = ["java", "scala", "python"]
upper = [name.upper() for name in lower]


lower, upper
```

```python
# Ü: startliste mit 10 Zufallszahlen im Bereich 20 - 50, Comprehension liefert: Liste der Quadrate
# starting list with 10 random numbers in the range 20 - 50, comprehension provides: list of squares
#import random

startliste = (random.sample(range(20, 51), 10))


startliste1 = [elem ** 2 for elem in startliste]
startliste, startliste1
```

```python
# Anton
import random

startliste = random.sample(range(20, 51), k=10)
ergebnisliste = [n ** 2 for n in startliste]

startliste, ergebnisliste
```

```python
list_mit_duplikaten = [1, 3, 6, 3, 1, 7]
menge_ohne_duplikate = {elem for elem in list_mit_duplikaten}

list_mit_duplikaten, menge_ohne_duplikate
```

```python
{n*n for n in [1, 5, 7]}
```

```python
[n * 2 for n in range(1,10)]
```

```python
[n * 2 for n in range(1,10) if n % 2 == 1]
```

```python
"""
This is an example of a dictionary comprehension
"""
# {}  schleifen klammern

{n: n * 10 for n in range(5)}
```

```python
from datetime import datetime, timedelta

zeitpunkt = datetime (2024, 10, 13, 17, 25)
abstand1 = timedelta(hours=2, minutes=10)
abstand2 = timedelta(minutes=3, seconds=28)

zeitpunkt_zukunft = zeitpunkt + abstand1
zeitpunkt_vergangenheit = zeitpunkt - abstand2

termine = [zeitpunkt_vergangenheit, zeitpunkt, zeitpunkt_zukunft]


termine
```

```python
delta = timedelta(minutes=10)
# jedestermin
[termin + delta for termin in termine]
```
