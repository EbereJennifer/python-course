#### Coordinated Universal Time - UTC
UTC is the global reference time.
It does not change with daylight saving.
Think of it as the “clock of the world” that everyone can convert to. e.g
   `UTC time: 2024-04-09 20:00`
#### pacific standard time - PST
- PST is UTC−8 hours.
- This means PST is 8 hours behind UTC.
####

```python
from datetime import timezone, timedelta, datetime

pst = timezone(timedelta(hours=-8))
dt = datetime(2017, 8, 1, 15, 41, tzinfo=pst)  # tzinfo → attaches timezone to datetime

dt
```

```python
berlin_tz = timezone(timedelta(hours=2))
new_york_tz = timezone(timedelta(hours=-4))

zeitstempel_berlin = datetime(2026, 4, 9, 11, 56).replace(tzinfo=berlin_tz)
zeitstempel_new_york = zeitstempel_berlin.astimezone(new_york_tz)

zeitstempel_berlin, zeitstempel_new_york
```

```python
from zoneinfo import ZoneInfo

zone_berlin = ZoneInfo("Europe/berlin")
zone_berlin

dt1 = datetime(2020, 4, 18, 9, 30, tzinfo=zone_berlin)
dt1
```

```python
dt1 = datetime(2026, 3, 29, 1, 59, tzinfo=zone_berlin)
dt2 = datetime(2026, 3, 29, 3, 13, tzinfo=zone_berlin)
dt1

delta = timedelta(minutes=15)
dt1, dt1 + delta, dt2 - dt1, dt2.isoformat()
```

```python
zone_utc = ZoneInfo("UTC")
utc1 = dt1.astimezone(zone_utc)
utc2 = dt2.astimezone(zone_utc)


# Empfehlung: Rechnungen mit Sommer/ Winterzeit: immer vorher nach UTC umwandeln
utc2 - utc1
```
