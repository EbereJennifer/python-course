```python
# beliebig viele keyword argument mit **

def f(**kwargs):
    print(kwargs)
f()
#f(1) this will give a TypeError

f(a=1, b=2, c=3)
# f(2, a=1) this will give an error
f(x=1, a=3, h=7)
```

```python
def f(n, **kwargs):
    print(n, kwargs)


f(3)
f(3, k=6)       # every first word or arguments, takes the position of n parameter. n can either be specified or not
f(n=1, x=7)
f(x=6, n=1)
```

```python
def f(*args):
    print(args)   # none is the result from print(args)


f()
f(1)
f(1, 2, 3,)
# f(a=1) benannt ist nicht erlaubt
```

```python
# alle parameter nach die sternchen *, brauchen Namen
def f(a, *, b):
    print(a, b)


f(1, b=2)
# f(1, 2) verboten
f(b=2, a=1)
f(a=23, b=564)
```

This code demonstrates a special kind of function parameter in Python called a positional-only parameter.

def f(a, /, b):
    print(a, b)

The key part is the /.

Everything before / → positional-only parameters
Everything after / → can be positional OR keyword

So here:

a → positional-only
b → flexible (positional or keyword)
🔹 What “positional-only” means

A positional-only parameter:

Must be passed by position
Cannot be passed using its name
🔹 Valid Calls
f(1, 2)
a = 1 (position 1)
b = 2 (position 2)
✔ Works
f(1, b=2)
a = 1 (positional)
b = 2 (keyword)
✔ Works
🔹 Invalid Calls
f(a=1, b=2)

❌ Error because:

a is positional-only, but you're trying to pass it as a keyword
f(b=2, 1)

❌ Syntax error because:

Positional arguments must come before keyword arguments
🔹 Why this feature exists

The / syntax was introduced in Python 3.8. It’s useful when:

You don’t want users to depend on parameter names
You can later rename a without breaking code
You want a cleaner or stricter API
Common in built-in functions (e.g., len(obj) doesn’t allow len(obj=...))

```python
# Ü: In 3.8 Neuerung: Parameter links von / nur positional
#
def f(a, /, b):
    print(a,b)


f(1, 2)    # a must not be specified with "="
f(1, b=2)
# f(a=1, b=2) verboten
# f(b=2, 1) verboten
```

```python
# you can pass all arguments by position only
def f(a, b, c, d, /):
    print(a, b, c, d)


f(1, 2, 3, 4)

# f(a=1, b=2, c=3, d=4) verboten
```

```python
# Zweiter Parameter ist optional, arguments can be placed in any position but the result always comes in the position of the print statement
def f(a, b=2):
    print(a, b)

f(1)
f(1, 3)
f(1, 2)
f(a=1, b=2)
f(b=18, a=1)
```
