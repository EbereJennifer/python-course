```python
import random
import csv
from datetime import date, timedelta

daten = {}

for i in range(1, 11):
    datum = date.today() - timedelta(days=i)
    werte = []
    for _ in range(4):
        wert = random.gauss(9.0, 4.0)
        werte.append(wert)
    print(datum, werte)
    daten[datum] = werte

with open("daten.csv", mode="w", newline="") as csv_file:
    writer = csv.writer(csv_file, delimiter=";")
    for item in daten.items():
        list_for_row = []
        list_for_row.append(date.isoformat(item[0]))
        for temp in item[1]:
            list_for_row.append(temp)
        writer.writerow(list_for_row)


print("*" * 100)


daten = {}

print("Einlesen")

with open("daten.csv", mode="r", newline="") as csv_file:
    reader = csv.reader(csv_file, delimiter=";")
    for row in reader:
        string_datum = row[0]
        datum = date.fromisoformat(string_datum)
        werte = row[1:]
        daten[datum] = [float(wert) for wert in werte]

for datum in daten.keys():
    print(f"{datum}: {daten[datum]}")
```

```python
# Anton
# 1

class NoDataException1(Exception):
    pass

def show_statistics(zahlenliste):
    if not zahlenliste:
        raise NoDataException1("Liste darf nicht leer sein")

    min = zahlenliste[0]
    max = zahlenliste[0]
    summe = zahlenliste[0]

    for i in range(1, len(zahlenliste)):
        elem = zahlenliste[i]
        summe += elem
        if elem < min:
            min = elem
        if elem > max:
            max = elem

    durchschnitt = summe / len(zahlenliste)
    print(f"Minimum: {min}")
    print(f"Maximum: {max}")
    print(f"Durchschnitt: {durchschnitt}")
    print(f"Anzahl: {len(zahlenliste)}")

    return (min, durchschnitt, max, len(zahlenliste))

show_statistics([4, 7, 8, 2, -1])
try:
    show_statistics([])
except NoDataException1 as ex:
    print("********* NoDataException1 wurde geworfen", ex)
# print("Denach")
```
