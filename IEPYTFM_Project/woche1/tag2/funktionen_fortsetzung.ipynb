##### Vorbetrachtung: Entpacken
Eine Liste oder ein Tuple kann mehrere Variablen verteilt werden: A list or a tuple can be distributed among several variables

```python
a, b, c = [23, 54, -12] # Anzahl muss übereinstimmen (match)
# a, b, c stehen als einzelne Variable bereit. are available as individual values
print(a)

d, e, f, g = (True, False, True, False)
f
```

Das Entpacken geht auch beim Funktionsaufruf . mit Liste oder Tuple können mehrere Argumente übergeben werden. Notation: * vor Liste oder Tuple(spread). Unpacking is also possible when calling a function. Multiple arguments can be passed using a list or tuple. Notation: * before the list or tuple (spread)

Die Anzahl muss übereinstimmen:
- Parameter innerhalb der Funktion
- Argumente, die der Funktion von außen übergeben werden
- Arguments passed to the function from outside

```python
def aufsummieren(a, b, c, d):       # the * symbol is called the unpacking operator (or “argument unpacking”).
    return a + b + c + d            # It takes a list (or any iterable) and unpacks its elements into separate arguments                                # when calling a function.
aufsummieren (*[2, 7, 8, 5])        # 5 Argumente  will give an error without *
```

```python
# wäre umständlicher ohne * Operator:  It would be more cumbersome without the * operator:

liste = [4.5, -0.9, 12.1, 7.5]
# aufsummieren(*liste)
aufsummieren(liste[0], liste[1], liste[2], liste[3])
```

```python
aufsummieren(*liste)
```

#### Aufruf über Position oder über Parameternamen
"Normale" Parameter können entweder über ihre Position oder über ihre Namen übergeben werden: bei der zweiten Varianten kann die Reihenfolge beim Aufruf verändert werden. “Normal” parameters can be passed either by their position or by their names; in the latter case, the order can be changed when the function is called.


```python
def print_two_strings(s1, s2):
    print(s1, s2)

print_two_strings("Hallo", "Python")
print_two_strings(s1="Hallo", s2="Python")
print_two_strings(s2="Anaconda", s1="Hallo")
print_two_strings(s2="Numpy", s1="Hallo")
# print_two_strings(s1="Hallo", "Numpy")   SyntaxError: positional argument follows keyword argument, nicht erlaubt
```

#### schlüsselwortparameter (key word Parameter)
Man kann Parameter einrichten, die nur per Name übergeben werden können. Diese werden in der Parameterliste durch * von der vorigen abgetrennt. Möglich: Nur solche Schlüsselwortparameter. You can set up parameters that can only be passed by name. These are separated in the parameter list from the previous ones by *. Possible: Only such keyword parameters

```python
def misch_funktion(a, *, b, c):    # The * means: All parameters after * must be given as keyword (named) arguments

    print(f"a={a}, b={b}, c={c}")

misch_funktion(3, b=6, c=5)       # notice that "a" wasn't given here, but the answer shows a=3
misch_funktion(a=1, b=7, c=2)
misch_funktion(b=4, c=9, a=-1)
misch_funktion(a=5, b=8, c=3)

# a → can be passed positionally OR by name
# b and c → must be passed by name
```

#### Standardwerte
Parameter in Funktionen können Standardwerte erhalten, diese müssen hinten stehen. Auch genannt: Optionale Parameter.
Parameters in functions can have default values, which must be listed last. Also known as: optional parameters

```python
def begrüßung(name, floskel="Hallo"):   # phrase=“Hello”
    print(f"{floskel}, {name}!")

begrüßung("Steven", "Guten Tag")
begrüßung("Maria")


def ausgabe_reduzierte_preise(standard_preise, prozent_reduzierung=10.0):

    for preis in standard_preise:
        reduziert = preis * (100.0 - prozent_reduzierung) / 100.0
        print(f"Standard: {preis}, reduzierter preis: {reduziert:.2f}")


ausgabe_reduzierte_preise([9.99, 20, 14.89], 10.0)   # further reduce to 15%


```

Eine Funktion kann mit dem Schlüsselwort return einen Wert (Ergebnis) zurückgeben
. Alle Typen sind hierfür möglich. A function can return a value (result) using the keyword `return`.
Any type can be returned.

#### Variable Anzahl von Parametern
Durch *args in der parameterliste kann eine belibige Anzahl von Argumenten übergeben werden. The *args in the parameter list allows any number of arguments to be passed

```python
def fun_beliebig_viele_argumente(*args):
    print(type(args))
    print(f"Anzahl übergebener Argumente: {len(args)}")

    for arg in args:
        print(arg, end=" ")
    print("\n")

fun_beliebig_viele_argumente(1, 2, 3, 8, -6, 5)
fun_beliebig_viele_argumente(True)
fun_beliebig_viele_argumente(0)
fun_beliebig_viele_argumente()          # this an empty tuple
```
