Markup languages are systems for annotating text so that a computer can understand how the text should be structured, displayed, or processed.
A markup language adds tags or symbols around content to describe what it is.

# Example (HTML): is the style used in websites
- <h1>This is a heading</h1>
- <p>This is a paragraph.</p>
- <h1> → tells the browser this is a heading
- <p> → tells it this is a paragraph*



Key characteristics
- Uses tags or markers
- Focuses on structure and formatting
- Doesn’t contain logic like loops or conditions (in most cases)

Common markup languages
- HTML → for web pages
- XML → for storing and transporting data
- Markdown → for simple formatting (like # Heading)

Markup Sprachen können strukturierte Texte mit Auszeichnungen (Markierungen) erstellen.

Mischungen sind möglich (Latex innerhalb Markdown)

Markup languages can create structured text using markup.
Examples:
- Tex: extremely complex
- LaTeX: complex
- XML: versatile
- HTML: used for websites
- JSON: often used as a simple alternative to XML; flexible
- Markdown: relatively simple; mostly used for documentation

Mixing markup languages is possible (e.g., LaTeX within Markdown)
