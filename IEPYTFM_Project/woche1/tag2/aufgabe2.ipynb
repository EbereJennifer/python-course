#### Enumerate()
enumerate() lets you loop over a list and keep a counter, start counting at 1, instead of the default 0. So enumerate(argumentlist, 1) produces pairs like:

- (1, (2, 3, 4))
- (2, (12, 43))
- (3, (578,))
- (4, ())
#### for *i*, args in....

- This unpacks each pair into two variables:

- i → the counter (1, 2, 3, 4)
- args → each tuple from the list


```python

def berechne_v(width=1, length=1, height=1):
    info = f"width: {width}, length: {length}, height: {height}"
    print(f"Parameter value.", info)
    return width * length * height

argumentlist = [
    (2, 3, 4),
    (12, 43),
    (578,),
    ()
]

for i, args in enumerate(argumentlist, 1):
    print(f"{i}. Durchlauf:")
    vol = berechne_v(*args)
    print(vol)

# Ü: Mit den Argument values aus der argumentliste funktion aufrufen,
# auch ausgeben: 1. Dürchlauf usw
```

3) Ein Warenkorb soll als eine Liste von Tupel der Form (produkt_name, einzel_preis, anzahl) aufgebaut werden.
Funktion berechne_gesamtpreis(warenkorb) schreiben.


```python
warenkorb = [
    ("car", 2750.78, 3),
    ("Laptop", 1480.67, 16),
    ("printer", 689.56, 56)
]
def berechne_gesamtpreis(warenkorb):
    total = 0                       # This variable will store the final sum
    for product_name, unit_price, quantity in warenkorb:
        total += unit_price * quantity
    return total


print(f"The berechne gesamtpreis ist {berechne_gesamtpreis(warenkorb):.1f} Euro")
```
