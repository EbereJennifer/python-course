### Funktionsobjekte

Wenn man eine Funktion definiert, wird ein Funktionsobjekt angelegt. Funktion funktioniert intern wie
andere Objekte, hat z.B. interne Daten und evt. Methoden.

Funktionen sind "Objekte erster Klasse" und können z.B. als Argument übergeben oder als Returnwert
zurückgegeben werden.

When you define a function, a function object is created. Internally, a function behaves like
other objects; for example, it has internal data and may have methods.

Functions are “first-class objects” and can, for example, be passed as arguments or returned as return values.

Examples of functions used as parameters:
Beispiel für Funktion als Parameter:

- filter bei Listen
- map bei Listen
- sort bei Listen

| Function        | What it does     | Function parameter role |
| --------------- | ---------------- | ----------------------- |
| `filter()`      | selects items    | condition (True/False)  |
| `map()`         | transforms items | transformation          |
| `sort(key=...)` | defines order    | sorting rule            |






#### filter() with lists

- filter() takes:
    - a function (condition)
    - a list
- It keeps only the elements where the function returns True.

```python
def is_even(x):
    return x % 2 == 0

numbers = [1, 2, 3, 4, 5]

result = filter(is_even, numbers)
print(list(result))
```

#### map() with lists

- map() takes:
    - a function (transformation)
    - a list
- It applies the function to every element.

```python
def square(x):
    return x * x

numbers = [1, 2, 3]

result = map(square, numbers)
print(list(result))
```

sort() with lists

👉 sort() can take a function that defines how to sort

```python
words = ["apple", "banana", "kiwi"]

words.sort(key=len)             # len is passed as a function
                                # Sorting is based on word length
print(words)                    # len → passing the function
                                # len() → calling the function immediately

```

- When editing, tabs are automatically converted to 4 spaces.
- Recommended: 2 spaces after a function definition, this means, leave 2 blank lines before and after top-level functions
- Function names consist of letters, numbers, and underscores; do not use predefined names (list, set, dict, etc.)

- Summary
- Use 4 spaces for indentation
- Avoid tabs (or let editor convert them)
-  Leave blank lines to make code readable:

#### Funktionen mit Parameter
Beliebig viele Parameter sind möglich. Beim Aufruf müssen entsprechend viele Werte übergeben werden. di Funktion kann dann arbeiten (auswerten, ausgeben, weiterrechnen...
Any number of parameters is possible. When the function is called, the corresponding number of values must be passed to it. The function can then perform its task (evaluate, output, continue calculating...).

```python
# Ü: Funktion quadrat mit 1 parameter, Funktion addition mit zwei Parametern
# definieren und ausführen

# Here, zahl refers to a parameter that has a name
def quadrat(zahl):
    return zahl * zahl


def addition(a, b):
    return a + b


# man muss genau die richtige Anzahl an Argumenten übergeben
print(quadrat(7))         # aufruf mit one argument for the zahl parameter
print(addition(3,1))      # aufruf with 2 arguments

addition(3, 5)
# addition(3, 5, 9)        this will give an error
```
