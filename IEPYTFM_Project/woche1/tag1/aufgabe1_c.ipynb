```python
# 2) Funktion print_in_spalten_version_a soll zwei Parameter erhalten: Eine Liste mit Strings sowie eine
# Liste mit den gewünschten Spaltenbreiten. Die Strings der Liste sollen rechtsbündig in den Spalten
# ausgegeben werden, alles in einer Zeile. The function `print_in_columns_version_a` should take two parameters: a list of strings and a list of the desired column widths. The strings in the list should be output right-aligned within the columns, all on a single line.


def print_in_spalten_version_a(strings, breiten):   # strings and breiten are first and second parameters
    for i in range(len(strings)):
        string = strings[i]        # [i] is the index position of strings e.g index 0 (first item) 1(second item) 2(third item)....                           # so this means, give me the elements at position [i] and store it in a variable called string
        breite = breiten[i]
        print(f"{string:>{breite}}", end=" ")


string_liste1 = ["Java", "Golang", "Ruby", "Javascript", "Objective C"]
string_liste2 = ["C", "C++", "C#", "kotlin", "Visual C++"]
breiten_liste = [10, 10, 10, 20, 20]
print_in_spalten_version_a(string_liste1, breiten_liste)           # string_liste1 and breiten_liste are called objects
print()
print_in_spalten_version_a(string_liste2, breiten_liste)
```

```python
# 3) Funktion print_in_spalten_version_b soll dieselbe Funktionalität wie print_in_spalten_version_a haben. Sie soll aber nur einen einzigen Parameter (ein Dictionary) haben.

# my own solution : Ebere

def print_in_spalten_version_b(strings):
    for i in range(len(strings)):
        string = strings[i]
        breite = len(string)
        print(f"{string:>{breite}}", end=" ")


string_liste1 = ["Java", "Golang", "Ruby", "Javascript", "Objective C"]

print_in_spalten_version_b(string_liste1)
```

```python
# 3) Funktion print_in_spalten_version_b soll dieselbe Funktionalität wie print_in_spalten_version_a haben. Sie soll aber nur einen einzigen Parameter (ein Dictionary) haben.

# correct version : Anton

def print_in_spalten_version_b(strings_to_breiten):
    for string in strings_to_breiten:
        breite = strings_to_breiten[string]
        print(f"{string:>{breite}}", end=" ")

string_liste = ["Java", "Golang", "Ruby", "Javascript", "Objective C"]
breiten_liste = [10, 10, 10, 20, 20]
dictionary = dict(zip(string_liste, breiten_liste))
print_in_spalten_version_b(dictionary)
```

##### Weitere Variante als Übung
Gewünscht: Gleicher Output, print_in_spalten_version_c hat als Parameter eine Liste mit 2er_tuple z.B [("Java", 10]


```python
def print_in_spalten_version_c(liste_mit_tuple):
    for pair in liste_mit_tuple:
        string = pair[0]            # pair[0] → the text (e.g., "Rot")
                                    # air[1] → the width (e.g., 1)
        width = pair[1]
        print(f"{string:>{width}}", end=" ")

tuple_list = [
    ("Rot", 5),
    ("Violet", 10),
    ("Gelb", 7),
    ("Schwarz", 10),
    ("Rot", 5)
]

print_in_spalten_version_c(tuple_list)
```
