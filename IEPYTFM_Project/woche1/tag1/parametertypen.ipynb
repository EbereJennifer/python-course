```python
# Funktion mit Liste als Parameter
def give_me_every_second_element(liste):
    result = []
    for i, elem in enumerate(liste, 1):
        if i % 2 == 0:
            result.append(elem)
    return result

liste = ["Berlin", "Munchen", "Koln", "Hamburg", "Leipzig"]
print(give_me_every_second_element(liste))

# print(give_me_every_second_element(True))   # TypeError: 'bool' object is not iterable

try:
    give_me_every_second_element(True)
except TypeError:    # dynamic typing
    print("Ungeeigneter Typ des Parameters")
```

```python
# Ü: Funktion build_dict_from_set_and_list mit zwei Parametern set_keys, list_values

def build_dict_from_set_and_list(set_keys, list_values):

    result = {}
    for key, value in zip(set_keys, list_values):
        result[key] = value
    return result


nummern = set(range(5))
einheiten = ["km", "m", "cm", "mm", "kg", "g", "h", "min"]

build_dict_from_set_and_list(nummern, einheiten)
```
