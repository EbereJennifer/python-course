#### while-loop

Key idea

- A while loop needs 3 things:

* Start value: i = 0 . i value at 0

* Condition: while i < 6:

* Change (update): i += 1

* Without the update step, the loop never ends.

#### while-loop(break)
break immediately exits the loop, no matter what the condition says.

```python
for x in range(1,11):
    print(x, end= " ")
# while loop

i = 0
while i < 6:
    print(i)
    i += 1    # This increases i by 1 each time the loop runs.
"""
# i = 0
# while i < 6:
#     print(i)  this will give an infinite loop because 0 < 6 is always true
"""
i = 0

```

```python
while i < 10:
    if i == 5:
        break
    print(i)
    i += 1
print(i, end=" ")
```
