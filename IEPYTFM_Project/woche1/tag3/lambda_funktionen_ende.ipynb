#### Rekursion
Recursion is a fundamental concept in computer science where a function solves a problem by calling itself on smaller, simpler versions of the same problem.
- Base case – the condition where the function stops calling itself.
- Recursive case – the part where the function calls itself with a reduced input.
- Technik in der Programmierung für Funktionen
- Funktion kann sich selbst (mit veränderten Argumentwerten) aufrufen
- Abbruch bei einem Basisfall
- A programming technique used for functions
- A function can call itself (with different argument values)
- Termination occurs in the base case

```python
def fakultaet(n: int) -> int:
    if n == 1:              # basisfall
        return 1
    else:
        return n * fakultaet(n - 1)  # if you use only "n" here, the loop continues to run and never stops

for n in range(2, 20):
    print(fakultaet(n))
```
