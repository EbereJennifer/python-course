The recursive function `quersumme_rekursiv(zahl)` and the iterative function `quersumme_iterativ(zahl)` using a loop; both are intended to calculate the sum of the digits, e.g., the sum of the digits of 143 is 1 + 4 + 3 = 8. (quersumme = sum of digits)

#### Quersumme: means digit sum

    if zahl < 10:

    else:
        return zahl % 10 + quersumme_rekursiv(zahl // 10) # That doesn’t return 26 — it calls the function again with 26.

quersumme_rekursiv(2)

````python
def quersumme_rekursiv(zahl: int) -> int:
    if zahl < 10:
        return zahl
    else:
        return zahl % 10 + quersumme_rekursiv(zahl // 10)
print(quersumme_rekursiv(268))
````

````python
# iterative means it must be in a loop

def quersumme_iterativ(zahl: int) -> int:
    sum = 0
    while zahl > 0:
        sum += zahl % 10
        zahl = zahl // 10
        return sum
print(quersumme_rekursiv(269))
````

````python
#  Anton

# Aufgabe 1

def quersumme_rekursive(zahl):
    assert zahl > 0, f"input {zahl} nicht erlaubt, muss > 0 sein"

    if zahl < 10:
        return zahl
    else:
        rest = zahl % 10
        ganzzahlig_enthalten = zahl // 10
        return rest + quersumme_rekursive(ganzzahlig_enthalten)


liste_input = [1234, 3445, 897687, 985646783]
liste_ergebnisse = list(map(lambda zahl: quersumme_rekursive(zahl), liste_input))

for input, quersumme in zip(liste_input, liste_ergebnisse):
    print(f"Input: {input}, Quersumme: {quersumme}")
````

````python
# Aufgabe 2

# The function `rekursiv_min_max` should calculate the minimum and maximum of a list of integers using recursion. It should return a tuple as the result.
# Use type hints and use `assert` to check that the list is not empty.
#
# Example: `rekursiv_min_max([3, 7, 5, 6, 1, -3, 4])` should return `(-3, 7)`
#
# Possible helpful preparation: Calculate only the minimum as a recursive function.

# Evt. hilfreich als Vorbereitung: Nur das Minimum als rekursive Funktion berechnen.



def minimum_rekursiv(liste):
    if (len(liste) == 1):
        return liste[0]
    else: # mindestens 2 Elemente
        slice = liste[1:] # hinterer Teil ohne Startelement
        minimum_slice = minimum_rekursiv(slice)
        start_element = liste[0]
        if start_element < minimum_slice:
            return start_element
        else:
            return minimum_slice

liste = [-20, 2, 6, -1, 7, 9, -11]
ergebnis = minimum_rekursiv(liste)
print(ergebnis)


def rekursiv_min_max(liste):
    assert len(liste) > 0
    first = liste[0]

    if len(liste) == 1:
        return (first, first)
    else:
        slice = liste[1:]
        (min, max) = rekursiv_min_max(slice)
        if first < min:
            min = first
        if first > max:
            max = first
        return (min, max)


erg = rekursiv_min_max([2, -5, -3, 5, 11, -2])
print(erg)
````

````python
# Anton
# Aufgabe 3




def multiplikation(a: int, b: int) -> int:
    assert a >= 0

    if a == 0: # Basisfall
        return 0
    else: # a > 0 rekursiver Aufruf mit kleinerem erstem Parameter
        return multiplikation(a - 1, b) + b


print(multiplikation(9, -3))

# Vorüberlegung: 3 ** 4 == 3 * 3 * 3 * 3
def potenz_funktion_rekursiv(a: int, b: int) -> int:
    assert b >= 0

    if b == 0:
        return 1
    else: # Ü fertigmachen
        teilergebnis = potenz_funktion_rekursiv(a, b - 1)

        return a * teilergebnis
print(potenz_funktion_rekursiv(2, 10))


````

````python
# . Iterative version (fastest & simplest)
def f(n):
    if n < 3:
        return n
    a, b = 1, 2  # f(1), f(2)
    for _ in range(3, n + 1):
        a, b = b, a + b
    return b

print(f(4))
````

```let’s walk through it:

f(5)
→ f(4)
→ f(3)
→ f(2)
→ f(1)
→ f(0)
→ 0

Now the returns go back up the call stack:

f(1) returns f(0) → 0
f(2) returns f(1) → 0
f(3) returns f(2) → 0
f(4) returns f(3) → 0
f(5) returns f(4) → 0
Key idea:

There’s no calculation happening — each call just passes along the result from the next one.

So the function is basically saying:

“Keep going down until 0, then return 0 all the way back up.”

````python
def f(n):
    if n == 0:
        return 0
    return f(n-1)

print(f(5))
````
