```python
# go to the command line and write "pip install matplotlib"
# pip show pandas (no pandas for now)

import matplotlib.pyplot as plt
# figure: ganze Grafik, axes<. x- und y-Achse

fig, ax = plt.subplots()
# fig → the Figure
# ax → the Axes

x_werte = ["A", "B", "C", "D"]
y_werte = [5, 7, 12, 4]

ax.plot(x_werte, y_werte)

ax.set_xlabel("Buchstabe")
ax.set_ylabel("Zahlenwert")
ax.set_title("Beispieldiagramm")

plt.show() #bei Jupyter nicht benötigt
```

```python
_, ax = plt.subplots()
monate = ["Januar", "Februar", "Marz", "April"]
temp_berlin = [2.3, -1.8, 4.7, 10.4]
temp_Munchen = [-1.4, -3.2, 5.4, 9.8]

ax.plot(monate, temp_berlin)
ax.plot(monate, temp_Munchen)
ax.set_xlabel("Monate")
ax.set_ylabel("Grad Celsius")
ax.set_title("Wetter in Berlin und Munchen")

plt.show()
```

```python
# or

_, ax = plt.subplots()
monate = ["Januar", "Februar", "Marz", "April"]
temp_berlin = [2.3, -1.8, 4.7, 10.4]
temp_Munchen = [-1.4, -3.2, 5.4, 9.8]

ax.plot(
    monate,
    temp_berlin,
    label = "Berlin",
    color = "green",
    linestyle = "dashed",
    marker = "o",
)

ax.plot(
    monate,
    temp_Munchen,
    label = "Munchen",
    color = "blue",
    linestyle = "dashed",
    marker = "o",
)

ax.legend()

ax.set_xlabel("Monate")
ax.set_ylabel("Grad Celsius")
ax.set_title("Wetter in Berlin und Munchen")

plt.show()


```
