This is about modularization (German: Modularisierung) in programming.

🔹 What it means

Modularization means: Splitting your code into smaller, separate parts (modules) instead of writing everything in one big file.

A module can be:

a separate Python file (.py)
a group of related functions/classes
🔹 Simple example

Instead of one big file:

def add(a, b):
    return a + b

def multiply(a, b):
    return a * b

print(add(2, 3))

👉 You split it:

math_utils.py

def add(a, b):
    return a + b

def multiply(a, b):
    return a * b

main.py

import math_utils

print(math_utils.add(2, 3))
🔹 The advantages (your notes explained)
✅ Wiederverwendbarkeit (Reusability)

You can reuse code in other programs:

import math_utils

No need to rewrite the same functions again.

✅ Wartbarkeit (Maintainability)

Code becomes easier to:

understand 👀
test 🧪
fix 🔧

👉 Instead of searching through one huge file, you know:

“math stuff is in math_utils”
“main logic is in main.py”
🔹 Real-world analogy 🧠

Think of it like organizing things:

❌ One giant messy drawer
✅ Multiple labeled boxes (modules)
🔹 In Python specifically
Every .py file is a module
You use import to bring it in

#### Globale Variablen
- Variable, die ausserhalb von Funktionen definiert werden
- Im ganzen Modul sichtbar
- Möglichst nur selten anwenden
- Änderung in Funktion: Schlüsselwort global notwendig


- Verwendung globaler Variablen kann zu unerwünschten Seiteneffekten führen, da sich überall geändert werden können.
- Zu viele globale Variable machen den code schwerer lesbar und wartbar.
- Mögliche Namenskonflikte lokalen: fehlermöglichkeit

#### Globale Namensräume
Beim Start eines Python Scripts wird ein globaler Namensraum erstellt: Enthält alle Definitionen auf der obersten Ebene des Scripts, also nicht innerhalb Funktion oder klasse.
Der globale Namensraum existiert, solange das script läuft. Die Namen des Globalen namensraums sind in gesamten Script zugänglich, wenn sie nicht durch lokale Namen verdeckt werden.

#### Scopes (Lebensdauer)
- Globaler scope: Ab Definition im ganzen Script verfügbar
- Enclosing scope (umgebender scope) bei Funktionsschachtelung, ab Definition in äußerer und innerer Funktion verfügbar
- Lokaler scope: Ab Definition nur in Funktion selbst verfügbar
- Built-in scope: Eingebaute Funktionen wie len, print, help, in allen Modulen verfügbar

#### LEGB: Suchreihenfolge bei Variablennamen
Erst lokal, dann enclosing, dann global, dann builtin
