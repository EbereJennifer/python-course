####  Bold
double ** symbol before and after a statement, **This is bold** makes the statement bold

####  Italics
- single * symbol before and after a statement or word, *This is italic* italizes it.

####  Bold + Italic

- Three *** symbol before and after a word or statement, italizes and makes it bold. ***This is bold and italic***

#### Unordered Lists
Unordered list (bullet points with hyphen -)

#### Ordered Lists (numbered)

3. Third item

#### Nested lists
