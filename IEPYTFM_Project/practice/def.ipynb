```python
# Give a function a default value, so that even when you do not supply the argument, it automatically applies it

def increment(number, by=1):  # b= 1 is the required parameter. all other parameters should come before it
    return number + by


print(increment(2))
print(increment(2, 7))
```

```python
# Unpacking: Passing a variable number of arguments to a function
def values(*numbers):
    return numbers


print(values(1, 2, 3, 8, 16))
```

```python
def values(*numbers):
    total = 1
    for i in numbers:
        #total = total * i
        total *= i
    return total


print(values(1, 2, 3, 8, 16))
```

```python
# How to pass keyword arguments using ** (save information as key value pairs i.e a dictionary)

def save_user_info(**user):
    print(user)                 # user here is called an Object

save_user_info(id=1, name="John", age=24, city="New York")

## we can get the various keys in this dictionary by using square brackets, and passing a string


def save_user_info(**user):
    print(user["id"])
    print(user["name"])

save_user_info(id=1, name="John", age=24, city="New York")
```

```python
# Scope: is the region of a code where a variable is defined

def fizz_buzz(input):
    if input % 3 == 0:
      return "Fizz"
    if input % 5 == 0:          # instead of writing "else:", put return on the else line (if input is divisible by 3,
      return "Buzz"                                   # return Fizz, else return Buzz)
    if (input % 3 == 0) and (input % 5 == 0):
      return "FizzBuzz"
    return input


fizz_buzz(5)
```

Funktion print_in_spalten_version_a soll zwei Parameter erhalten: Eine Liste mit Strings sowie eine Liste mit den gewünschten Spaltenbreiten.
Die Strings der Liste sollen rechtsbündig in den Spalten ausgegeben werden, alles in einer Zeile.


```python
def print_in_spalten_version_a(string, width):
    for i in range(len(string)):
        s = string[i]
        w = width[i]
        print(f"{s:>{w}}", end=" ")
print_in_spalten_version_a(["laptop", "mouse", "cable", "socket"], [6, 5, 5, 6])
```

```python
def print_in_spalten_version_b(list_in_dict):

    for pair in list_in_dict:
        s = pair [0]
        w = pair [1]
        print(f"{s:>{w}}", end=" ")

tuple_list = [
    ("Rot", 5),
    ("Violet", 10),
    ("Gelb", 7),
    ("Schwarz", 10),
    ("Rot", 5)
]

print_in_spalten_version_b(tuple_list)

```
