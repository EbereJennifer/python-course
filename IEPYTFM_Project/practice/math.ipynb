```python

import math
import matplotlib.pyplot as plt


def wertetafel(start, end, anzahl):
    abstand = (end - start) / (anzahl - 1)

    x_werte = []
    y_werte = []

    for i in range(anzahl):
        x = start + i * abstand
        y = math.exp(x)

        x_werte.append(x)
        y_werte.append(y)

    return x_werte, y_werte


# get values
x, y = wertetafel(1, 5, 50)

# plot
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("exp(x)")
plt.title("Graph of e^x")
plt.grid()

plt.show()
```
