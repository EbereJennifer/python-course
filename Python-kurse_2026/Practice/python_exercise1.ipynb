```
# 4. Conditions
# Ask for a number and print:
# “Positive”, “Negative”, or “Zero”

number = int(input("what is your number: "))

if number > 0:
    print(f"Positive: {number}")
elif number < 0:
    print(f"Negative: {number}")
elif number == 0:
    print(f"Zero: {number}")
else:
    print(f"Out of range: {number}")

```

```
# 8. Calculate the sum of numbers from 1 to 100
#You start with total equal to 0, This variable will store the sum.
#Each loop:
#add current number i
#store result back in total
# Step-by-step example (first few iterations)
#i	total before	total after
#1	0	1
#2	1	3
#3	3	6
#4	6	10
total = 0
for i in range(1, 101):
    total += i
    print(f"sum:", total)
```

```
# 9. Print a 5×5 multiplication table
#i goes through: 1,2,3,4,5
#Each value of i represents a row
#Each value of j represents a column

#Multiply current values: i * j
#end=" " means: print results on the same line, separated by space
for i in range(1,6):
    for j in range(1,6):
         print(i * j, end=" ")
    print()
```

```
#11. Simple calculator:
#Ask for 2 numbers
#Ask for operation (+, -, *)
#Print result

n1 = int(input("number1:"))
n2 = int(input("number2:"))
op = (input("addition or substraction or multiplication:" ))

if op == "+":
    result = n1 + n2
    print(f"The result from + is: {result}")
elif op == "-":
    result = n1 - n2
    print(f"The result from - is: {result}")
elif op == "*" :
    result = n1 * n2
    print(f"The result from * is: {result}")

else:
    print("Invalid operation")
```

```
# 12.
# Guessing game:
# Set a number (e.g. 7)
# Let user guess until correct
import random
n = random.randint(0,51)

if n == 7:
    print(f"The correct number is {n}")
else:
    print(f" {n}, you are almost there, keep on guessing")
```

```
#a better way to solving number 12
secret = 7

guess = int(input("Guess the number: "))

while guess != secret:
    guess = int(input("Try again: "))

print("Correct!")
```

```
#solving number 12 with random

import random
n = random.randint(4,11)
guess = int(input("guess the number (0-50):"))
while guess != n:
    guess = int(f"Try again, sorry: {guess}")
print("Correct!")

```

```
# 13. Thinking / logic

# Ask for a number n:
# Print all numbers from 1 to n
# Also print their sum


n = int(input("Enter a number: "))

total = 0
for i in range(1, n + 1):
    print(i)
    total += i

print("Sum:", total)
```
