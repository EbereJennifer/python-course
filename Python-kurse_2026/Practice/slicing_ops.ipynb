#### Strings
sequence[start:stop:step]

- Strings are immutable (cannot be modified
- However you can create new objects that are versions of the original one
- Variable name can only be bound to one object
- concatenation only works for strings. except you cast objects that are not strings to strings. e.g str(3)

- start → index where slicing begins (inclusive)
- stop → index where slicing ends (exclusive)
- step → how much to skip each time (optional)

 - [:n] → start → index n
  - [n:] → index n → end
    - [-n:] → last n items
        - [::step] → skipping
                - [::-1] → reversing

```python
nums = [0, 1, 2, 3, 4, 5]
print(nums[1:4])

print(nums[:3])   # from beginning to index 3
print(nums[3:])   # from index 3 to end
print(nums[:])   # entire list (copy)
print(nums[::2])
nums[-3:]
nums[::-1]
```

#### Input command
- To remove white spaces within a code just like we have on line 4 or to stop a code from repeating it on the last code, use ".join"
- .rstrip() can also be used to remove trailing white spaces
- int() can only convert strings that look like whole numbers, e.g.:

int("5")     # ✅ works
int("42")    # ✅ works
int("3.5")   # ❌ error

```python
input_verb = input("input verb: ")
print(f"I can {input_verb} better than you")
print(5 * input_verb)
print(5 * (input_verb + " "))
print(" ". join([input_verb] * 5))
```
