- Write a program that
- saves a secret number
- asks the user for a number guess
- prints whether the guess is too low, too high, or the same as the secret

```python
secret_number = 40
number_guessed = int(input("Enter your guessed number: "))
if number_guessed < secret_number:
    print(f"{number_guessed} is too low, try again")
elif number_guessed == secret_number:
    print(f"{number_guessed} is the same as secret number {secret_number}")
else:
    print(f"{number_guessed} is too high, try again")
print("Thanks!")
```
