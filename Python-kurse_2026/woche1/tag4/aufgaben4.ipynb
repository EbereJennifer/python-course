```python
# Aufgabe 3:
# Zufallszahl im Bereich 0-23 erzeugen. Fallunterscheidung:
# Wenn Zahl im Bereich 6-11: Ausgeben "Guten Morgen"
# Wenn Zahl im Bereich 12-17: Ausgeben "Guten Tag"
# Wenn Zahl im Bereich 18-22: Ausgeben "Guten Abend"
# Sonst: Ausgeben "Gute Nacht"


import random
random.randint(0,23)


zahl = random.randint(0,23)
if 6 <= zahl <= 11 :
    print(f"Guten Morgen: {zahl}")
elif 12 <= zahl <= 17:
    print(f"Guten Tag: {zahl}")
elif 18 <= zahl <= 22:
    print(f"Guten Abend: {zahl}")
else:
    print(f"Guten Nacht: {zahl}")
```

```python
# The random.gauss(mu, sigma) method in Python (from the built-in random module) is used to generate a random number following a Gaussian (normal) distribution.
#
# What it does:
# Returns a floating-point number.
# The number is drawn from a normal distribution defined by:
# mu → the mean (center of the distribution)
# sigma → the standard deviation (spread or width)
# In simple terms:
#
# It gives you random values that cluster around a central value (mu), with most values near the center and fewer as you move away—like a bell curve.

import random
for _ in range(10):
    zufall = random.gauss(3.0, 0.2)
    print(zufall)


liste = list("fromadistancetheworldlooksblue")

random.choices(liste, k=5)
```
