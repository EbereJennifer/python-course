```python
# map(function, iterable)

numbers = [1, 2, 3, 4]

# Function to square a number
def square(x):
    return x * x

# Using map
result = map(square, numbers)

# Convert to list to see the result
print(list(result))

liste = ["Python", "C", "Java"]
ergebnis = map(len, liste)
print(type(ergebnis), ergebnis)
# man kann über dars ergebnis iterieren oder Liste erzeugen
# kann nur einmal durchlaufen werden
for n in ergebnis:
    print(n, end=" ")
print()
#
liste_nach_map = list(ergebnis)
print(liste_nach_map)


```

```python
#Erzeugt Liste mit 10 zufallszahlen (0.0 -1.0)
# Durch map neu Liste mit Wahrheitswerten, ob Zahl > 0.7
import random


def fun(zahl):
    return zahl > 0.7
zahlen = []
for i in range(10):
    zahlen.append(random.random())
print(zahlen)

map_ergebnis = map(fun, zahlen)
liste_wahrheitswerte = list(map_ergebnis)
print(liste_wahrheitswerte)


```
