```python
# aufgabe 1
# the loop automtically ends when you input Key: ende

dictionary = {}
while True:
    key = input("key: ")
    if key.lower() == "ende":
        break
    value = input("value: ")
    dictionary[key] = value
print(dictionary)
```

```python
for (key, value) in dictionary.items():
    print(f"{key} ==> {value}")
```
