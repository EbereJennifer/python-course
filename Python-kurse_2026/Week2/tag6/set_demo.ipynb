```python
"""
Sets are mutable”

A set itself can be changed after it’s created.

That means you can:

Add elements
Remove elements
Clear the set

Example:

s = {1, 2, 3}
s.add(4)      # now {1, 2, 3, 4}
s.remove(2)   # now {1, 3, 4}

So the container (the set) is flexible → mutable.

🔹 “Elements are immutable”

The items inside a set must NOT change.

In Python, this means:

Elements must be immutable types (like int, str, tuple)
You cannot put mutable objects (like list, dict, or another set) inside a set

Example:

s = {1, "hello", (2, 3)}   # ✅ allowed

But:

s = {[1, 2], 3}   # ❌ error (lists are mutable)
"""
```

```python
# sets are mutable

#Erzeugen
# warnung: nicht set als Variablenname nehmen (do not use set as a variable name)

# Order looks “random”
# Sets are unordered, so the elements may appear in a different order than you entered them.
# Nothing prints in a script

set1 = set([3.8, 5.2, 1.8])
set2 = {2,4,6} # empfohlen, wenn man einzelne Element angibt (recommended when specifying individual elements)
set1, set2
```

```python
#sets doesn't print duplicates. it prints just one number
num = {1, 3, 6, 4, 3, 8, 9}
num
```

```python
# sets can be edited by adding a new value to it, however, it ignores the value if repeated or duplicated. it will print 5, only once
num.add(5)
num.add(5)
num
```

```python
# .remove()
# Removes an element from a set
# Raises an error if the element is NOT present

# s = {1, 2, 3}
# s.remove(2)   # works → {1, 3}
# s.remove(5)   # KeyError

# 🔹 .discard()
# Also removes an element from a set
# Does NOTHING if the element is NOT present (no error)
```

```python
# you can remove specified values from a set
num = {1, 3, 6, 4, 3, 8, 9}
 #if you try to remove values that isnt present in the set, you get a key error. elements have to be present in the set

num.remove(1)
num.discard(8)
num
```

```python
# you can remove specified values from a set
num = {1, 3, 6, 4, 3, 8, 9}
 #if you try to remove values that isn't present in the set, you get a key error. elements have to be present in the set

num.discard(12)

num
```

```python
 a = {1, 2, 3, 4}
 b = {4, 5, 6}
 schnittmenge = a & b                   # intersection
 vereinigungsmenge = a | b              # union
 differenzmenge = b - a                 # difference (“take all elements in b that are NOT in a”): set here is not symmetric (a - b != b -a)
 symmetrische_differenzmenge = a ^ b     # symmetric_difference_set (Elements that are in either set, but NOT in both. Think: “everything except the overlap”
 # (a-b) means take all element in a that are not in b

 schnittmenge, vereinigungsmenge, differenzmenge, symmetrische_differenzmenge,  a - b
```
