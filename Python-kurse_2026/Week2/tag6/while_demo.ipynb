```python
# n = 5
# if n < 10 # while n < 10 (so long as)
# 🧠 Key idea
#
# A while loop runs as long as the condition stays True.
# If you never change the variables in the condition, it can run forever.
# n starts at 5
# The condition n < 10 is True
# So it enters the loop and prints 5
# Then it goes back to check the condition again…
#
# ⚠️ But nothing ever changes n, so:
#
# n is still 5
# 5 < 10 is still True
# It prints 5 again
# This repeats forever

#✅ How to fix it
#You need to update n inside the loop, for example:

n = 5
while n < 10:
    print(n)
    n = n + 1 #  n += 1
```

```python
# 🧠 Why while True: is used
#
# while True: means:
# “Run this loop forever, until I explicitly stop it.”
# So the loop doesn’t depend on a condition at the top. Instead, you control when it stops using break.


import random

while True:
    n = random.randint(1,6)
    print(n)

    if n == 6:
        print("the loop ist 6")
        break
```

```python
# Here’s a version that counts how many rolls it takes to get a 6:

import random

count = 0

while True:
    n = random.randint(1, 6)
    count += 1
    print(n)

    if n == 6:
        print("the loop is 6")
        print("It took", count, "rolls to get a 6")
        break
# 🔍 What’s new
# count = 0 → starts a counter
# count += 1 → increases every time you roll
# When 6 appears, it prints how many tries it took
```

```python
# ❓ Could we avoid while True?
#
# Yes! You can write it using a condition:

import random

n = 0

while n != 6:
    n = random.randint(1, 6)
    print(n)

# This works because:
#
# The loop runs while n is not 6
# It automatically stops when n becomes 6
# ⚖️ When to use which
# ✅ Use while True:
# When the stopping condition happens inside the loop (like user input, random events, etc.)
# ✅ Use while condition:
# When you can clearly define the condition before each loop iteration

# 🔑 Simple takeaway
# while True: = “loop forever, I’ll decide when to stop”
# while condition: = “loop while this is true”
```

```python
# while loop is a continues loop
# used also when you do not know what to expect from the user for instance.
where = input("You are in the lost forest. Go left or right?")
while where == "right":
    where = input("You are in the lost forest. Go left or right?")
print("You got out of the lost forest!")
```
