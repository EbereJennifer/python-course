```python
zahl = 456  #when this number is too long, it ignores all the formula for the column width'

print(f"zahl: {zahl} Nach der Zahl")
print(f"zahl: {zahl:8} Nach der Zahl") # this means that the column had a width of 8. spaltenbreite 8 rechtsbündig
print(f"zahl: {zahl:>8} Nach der Zahl") #spaltenbreite 8, rechtsbündig
print(f"zahl: {zahl:<8} Nach der Zahl") #linksbündig
print(f"zahl: {zahl:^8} Nach der Zahl")   # zentriert
print(f"zahl: {zahl:.^8} Nach der Zahl")
print(f"zahl: {zahl:#<8} Nach der Zahl")
```

```python
value = 34.56789
print(f"value: {value} Dahinter")
print(f"value: {value:.2f} Dahinter")  # gerundet
print(f"value: {value:10.2f} Da hinter") #gerundet
print(f"value: {value:>10.2f} Da hinter")  #gerundet
print(f"value: {value:<10.2f} Da hinter")  #gerundet
print(f"value: {value:^10.2f} Da hinter")  #gerundet
print(f"value: {value:010.2f} Da hinter")
```

```python
text = " Python "
length = len(text) + 10
print(f"{text:*<{length}}")   #text before sign
print(f"{text:*>{length}}")   #text after sign
print(f"{text:*^{length}}")  #text in the middle
```
