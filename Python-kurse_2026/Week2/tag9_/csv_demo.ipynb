#### Write into a csv file
this code takes structured data (names, ages, jobs) and saves it into a file called `beispiel.csv`

##### Open a file
- `with open("biespiel.csv", mode="w") as csv_datei:`
- This means: Create (or overwrite) a file named "beispiel.csv"
- "w" = write mode

##### Why with?
It automatically: opens the file closes it safely afterward

##### Create a CSV writer
- `writer = csv.writer(csv_datei, delimiter=";")` : This prepares a tool that writes rows into the file.
- delimiter=";" means: values are separated by semicolons (;) instead of commas
- So a row will look like: Max;29;Ingenieur

##### Write all rows
`writer.writerows(daten)` : This writes all rows at once into the file.

Equivalent to:
for row in daten:
    writer.writerow(row)

- Print confirmation
print("CSV-Datei erfolgreich geschrieben") : Just tells the user:

- “CSV file successfully written”



```python
# Das ist ein Häüfiger Dateityp bei data science
import csv

daten = [
    ['Name', 'Alter', 'Beruf'],
    ['Max', 29, 'Ingenieur'],
    ['Sophie', 22, 'Lehrerin'],
    ['Tom', 35, 'Entwickler']
]
with open("biespiel.csv", mode="w") as csv_datei:
    writer = csv.writer(csv_datei, delimiter=";")
    writer.writerows(daten)
print("CSV-Datei erfolgreich geschrieben")
```

#### What does newline="" do?
with open("beispiel.csv", mode="w", newline="") as csv_datei, It tells Python:
- “Don’t automatically modify line breaks — let the csv module handle them.” because without newline="", you can get extra empty lines in your CSV file — especially on Windows.

❌ Without newline="":

Your file might look like this:

Name;Alter;Beruf

Max;29;Ingenieur

Sophie;22;Lehrerin

👉 Notice the blank lines between rows 😬

With newline="":
Name;Alter;Beruf
Max;29;Ingenieur
Sophie;22;Lehrerin

👉 Clean and correct 👍

```python
# or
with open("beispiel.csv", mode="w", newline="") as csv_datei:
    writer = csv.writer(csv_datei, delimiter=";")
    # auch verfügbar: writerow
    writer.writerows(daten)
print("CSV-Datei erfolgreich geschrieben")
```

#### Read CSV file
"r" tells Python: “Open this file so I can read data from it, not write to it.”

- Other common modes (for comparison)
    - "r" → read (file must exist)
    - "w" → write (creates/overwrites file)
    - "a" → append (adds to file)
- `with open("beispiel.csv", mode="r", newline="")` as file: Opens the CSV file for reading.
- *"with"* ensures it closes automatically
- newline="" → prevents line break issues (same idea as before)

##### Simple summary

- "r" = “I only want to read the file”
- csv.reader = turns file into rows (lists)
- loop = process each row
- formatting = make output pretty

```python
# CSV Datei lesen
# newline ="": in python übernimmt nicht das Behandeln des Zeilenendes. das ist schon in csv eingebaut
# delimiter muss das passende Trennzeichen in der CSV datei sein

# Reading a CSV File
# newline =“”: Python does not handle the line endings. This is already built into the CSV format
# delimiter must be the correct separator in the CSV file

with open("beispiel.csv", mode="r", newline="") as file:
    reader = csv.reader(file, delimiter=";")
    for zeile in reader:
        name = zeile[0]
        alter = zeile[1]
        beruf = zeile[2]
        print(f"{name:>10}, {alter:>6}, {beruf:>12}")
```
