```python
long_path = Path("ordner1/ordner2/ordner3/test.txt")
long_path.exists()
```

```python
long_path.parent.mkdir(parents=True, exist_ok=True)
long_path.parent.exists()
```

```python
# Bisher: open und später close
# Eleganter mit with Struktur, automatisch geschlossen
with open(long_path, "w") as datei:
    datei.write("Python")
    # close passiert automatisch'
```

```python
with open(long_path, "w") as datei:
    datei.write("Python\n")
    datei.write("Java\n")
    datei.write("Swift\n")
```

```python
with open(long_path, "r") as datei:
    for zeile in datei:
        text_ohne_zeilenumbruch = zeile.strip()
        print(text_ohne_zeilenumbruch)
```

```python
# Ü: Zusätzlich Zeilennummer ausgeben
# 1. Zeile: Python
with open(long_path, "r") as datei:
    for zeile in datei:
        text_ohne_zeilenumbruch = zeile.strip()
        print(text_ohne_zeilenumbruch)
```

```python
# Enumerate

with open(long_path, "r") as datei:
    for (nr, zeile) in enumerate(datei, start=1):
        text_ohne_zeilenumbruch = zeile.strip()
        print(f"Zeile {nr}: {text_ohne_zeilenumbruch}")

```
