```python
"""
Aufgabe 1:
Eine Funktion read_int_from_input mit Docstring schreiben, die einen ganzzahligen Wert einliest. Ungültige Eingaben (z.B. "abc") sollen nicht zum Abbruch führen, sondern es soll eine kurze Meldung angezeigt werden und das Einlesen soll solange versucht werden, bis ein gültiger int-Wert vorliegt.
"""
# Write a function named `read_int_from_input` with a docstring that reads an integer value. Invalid input (e.g., “abc”) should not cause the function to terminate;
# instead, a brief message should be displayed, and the function should continue attempting to read the value until a valid integer is obtained.
```

```python
def read_int_from_input():
    """
    Give an integer value

    Argument:
    int(input("give an integer value:"))

    if invalid integer is entered, continue trial until a valid integer is obtained
    Return:
        Valid integer given and the result is:

    """

    while True:
        try:
            return int(input("give an integer value:"))
            print(f"valid integer is: {read_int_from_input()}")
        except ValueError:
            print(f"invalid literal for int, try again")
        else:
            print(f"result is: {read_int_from_input()}")

print(f"valid integer is: {read_int_from_input()}")
```

```python
# Aufgabe 2:
# Analog Funktion read_float_from_input für Kommazahlen.
# Similar to the `read_float_from_input` function for floating-point numbers.

def read_int_from_input():
    """
    Input a float value

    Argument:

    float(input("input a float value:"))

    Return:
        Valid float given and the result is:
    else:
         break and continue trial until a valid float is obtained

    """


def read_int_from_input():
    while True:
        try:
            return float(input("give a float value: "))
        except ValueError:
            print("invalid float value, try again")
        else:
            print(f"valid float value is: {read_int_from_input()}")
print(read_int_from_input())
```

```python
# Aufgabe 3:
# Drei Angaben einlesen lassen: Produktname, Anzahl, Einzelpreis.
# (z.B. "USB-Stick", 3, 8.99)
# Daraus Gesamtpreis ausrechnen und formatiert ausgeben lassen.
```

```python
def add_products():
    product = {}
    while True:
        name = input("  Product name: ")
        if name == "*":
            break
        try:
          quantity = int(input("  Quantity: "))
          price = float(input("  Price: "))
        except ValueError:
            print("invalid integer or float type, try again")
            continue                                                    # "continue" means “Skip the rest of this loop iteration and start the next one.”)you Immediately go                                                                back to the top of the loop,  and ask the user again to retry
        product[name] = {"quantity": quantity, "price": price}
    return product

show_table = add_products()
print(show_table)

for name, data in show_table.items():
    print(f"{name}: {data['quantity']} {data['price']}")
```

```python
total_price = sum(item["quantity"] * item["price"] for item in show_table.values())
print(f"total price is {total_price}")
```

```python
def read_float_from_input(prompt):
    """
   Liest solange Input ein, bis eine gultige kommazahl vorliegt

    parameter:
    Eingabe prompt
    Rückgabe:
        Eingelesene float Zahl
    """
    while True:
        try:
            text = input("  Input aValueErr float: ")
            zahl = float(text)
            return zahl       # beendet auch die Schleife
        except:
            print(f"invalid float nummer, try again")
            continue
        print()
read_float_from_input(" float: ")
```
