```python
# this is a 3 dimensional tensor, vorstellen mit Höhe, länge und Breite

tensor_3d = np.random.random((2, 3, 4)) #3 dimensional
print(tensor_3d)       #3 roles, each in 2 dimensions - my own understanding
```

```python
array_top_as_vector = array_top.reshape((12, 1))
print(array_top_as_vector)
```
