```python
#Liste ist veränderbar (mutable)
#append() oe insert() is a method for lists used to add objects or elements to a list.
temperaturen = [4.2, 8.3, 7.9]   # extend(self, iterable, /) unbound builtins.list method, Extend list by appending elements from the iterable

temperaturen.append(2.9)           # list.append() takes exactly one argument
temperaturen.append(2.9, 4.0)      # this will give TypeError
temperaturen.append((2.9, 4.0))    # .append now considers the items in the bracket as a single item
print(temperaturen)


temperaturen.insert(1, -2.3)        # specifies the position where the element to be inserted will be positioned
print(temperaturen)

neu_temperaturen = [-1.5, 6.4]
temperaturen.extend(neu_temperaturen)
print(temperaturen)

help(list.extend)

# append() → adds as a single element
# extend() → adds each element separately
```

```python
# .remove
# Entfernen von Elementen. the element .remove, takes out only one element
cities = ["Berlin", "London", "Paris", "Rom", "Madrid"]
cities.remove("Paris")
print(cities)

cities = ["Berlin", "London", "Paris", "Paris", "Rom", "Madrid"] # if there are duplicates, remove() takes out the first
                                                                    # occurring element of the duplicate
cities.remove("Paris")
print(cities)

# pop() deletes the last element and works with indexing. when a number is stated, the index number chooses the element to be removed

cities.pop(0)
print(cities)

cities.pop() # .pop() this empty pop sign automatically removes the last element in the list.
print(cities)

# cities.pop(100) #muss gültig sein
cities.pop(1)
print(cities)

#Globale Funktion zum löschen
del cities[1]
print(cities)
```
