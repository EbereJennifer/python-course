```python
# Globale Funktion liefert sortierte kopie der liste (Global function returns a sorted version of the list
"""
sorted() sorts strings based on their Unicode/ASCII values, and uppercase letters come before lowercase letters.
"""
namen = ["stefan", "Ulysses", "Anna", "Kim"]
alphabetische_namen = sorted(namen)
n = sorted(namen, key=str.lower)
namen, alphabetische_namen, n
```

```python
namen = ["stefan", "Ulysses", "Anna", "Kim"]
alphabetische_namen = sorted(namen)
namen_von_hinten_sortiert = sorted(namen, reverse=True)
namen, alphabetische_namen, namen_von_hinten_sortiert
```

```python
namen = ["stefan", "Ulysses", "Anna", "Kim"]
alphabetische_namen = sorted(namen)
namen_nach_länger_sortiert = sorted(namen, key=len)
namen_von_hinten_sortiert = sorted(namen, reverse=True)
namen, alphabetische_namen, namen_von_hinten_sortiert, namen_nach_länger_sortiert
```

```python
"""
filtern
#benögtigt: eine Funktion, die auf die einzelnen Listenelemente angewendet wird und die bool zurückliefert
"""
#Required: a function that is applied to each list element and returns a boolean value

def ist_gerade(n):       # ist_gerade means even'
    return n % 2 == 0    # returns: True → even number and False → odd number

gefiltert = filter(ist_gerade, [2, 4, 5, 2, 8, 4])   # “Go through the list and keep only the elements where ist_gerade returns True.”
for elem in gefiltert:
    print(elem, end=" ")
```
