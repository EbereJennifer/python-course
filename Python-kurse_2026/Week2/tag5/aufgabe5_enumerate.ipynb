```python
# Aufgabe 1:
# 100 Zufallszahlen zwischen 0.0 und 1.0 erzeugen,
# daraus zwei Listen machen: Eine für die Zahlen <= 0.5, eine für die Zahlen > 0.5.
# Länge der beiden Listen ausgeben.

import random

from woche1.tag2.aufgabe1 import ergebnis

less_equal_05 = []
greater_05 = []

for _ in range(100):
    x = random.random()
    if x <= 0.5:
        less_equal_05.append(x)
    else:
        greater_05.append(x)

print("<= 0.5:", len(less_equal_05))
print("> 0.5:", len(greater_05))
```

```python
import random

# generate 100 random numbers
numbers = [random.random() for _ in range(100)]

# split into two lists
nkleinerer = [x for x in numbers if x <= 0.5]
ngroßer = [x for x in numbers if x > 0.5]

# print lengths
print("<= 0.5:", len(nkleinerer))
print("> 0.5:", len(ngroßer))
```

```python
# Aufgabe 3:
# Liste mit mehreren Lotto-Tipps erzeugen (pro Tipp jeweils 6 aufsteigende Zahlen im Bereich 1 bis 49). Optional: Zufallszahlen verwenden, recherchieren, wie man eine Liste sortiert.


import random

# how many lottery picks you want
num_picks = 5

picks = []

for _ in range(num_picks):
    pick = sorted(random.sample(range(1, 49), 6))
    picks.append(pick)

# output
for i, pick in enumerate(picks):
    print(f"Pick {i+1}: {pick}")
```
