```python
# dict nicht als variable namen verwenden (Do not use “dict” as a variable name)
# ist vordefiniert als Konstruktor für ein Dictionary
empty_dict = {}
type(empty_dict), empty_dict
```

```python
# Dictionary enthält Paare: key ==> value (The dictionary contains key-value pairs: key => value)
dict1 = {
    "München": 1500000,
    "Berlin": 3800000,
    "Hamburg": 1800000,
}
type(dict1), dict1
```

```python
# you can iterate or go through loop in dictionary
# for-Schleife liefert nur die enthaltenen Keys (The for-loop returns only the keys it contains)
for info in dict1:
    print(info)
```

```python
nur_die_keys = dict1.keys()
type(nur_die_keys), nur_die_keys
```

```python
#Ü: values aus dict1 rausholen
nur_die_values = dict1.values()
type(nur_die_values), nur_die_values
```

```python
"""
items() gives pairs: (key, value) → {'München': 1500000, 'Berlin': 3800000, 'Hamburg': 1800000})
"""
keys_und_values = dict1.items()
type(keys_und_values), keys_und_values
```

```python
for item in dict1.items():
    print(item)
```
