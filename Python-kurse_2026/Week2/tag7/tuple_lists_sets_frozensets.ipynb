```python
#Duplicates can be inserted into a list
list1 = [4, 6, 7, 9]
list1.append(6)
list1
```

```python
# immutable
# Removes duplicates (like a set)
# Unordered
# frozensets is safe to use because it prevents accidental changes
# it Can be used as dictionary keys

frozen = frozenset(list1) # you cannot add to a frozen list, frozen doesnt print duplicates
#frozen.add(67) geht nicht, da immutable
frozen
```
