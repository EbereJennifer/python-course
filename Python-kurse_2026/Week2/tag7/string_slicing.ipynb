```python
satz = "Python ist eine elegante Programmiersprache."
lezter_start = len(satz) - 3   #It prints all 3-letter substrings (chunks of length 3) from the sentence, moving one character at a time.
for start in range(0, lezter_start + 1): #Starts at index 0 , Ends at lezter_start, +1 is needed because range() excludes the end
    slice = satz[start : start + 3] # This grabs:3 characters starting at position start. eg :start = 0 → "Pyt", start = 1 → "yth", start = 2 → "tho"
    print(slice, end=" ")
```

```python
# “Take every third letter from the string, starting at the beginning.” start, stop and step. note that spaces are counted in slice

satz[0:len(satz) : 3]
```
