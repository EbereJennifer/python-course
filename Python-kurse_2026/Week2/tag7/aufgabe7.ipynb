```python
"""

Aufgabe 1:
Ratespiel: Erst eine Zufallszahl im Bereich 1-20 erzeugen.
In einer Schleife input abfragen und dann ausgeben, ob input zu klein, zu groß oder genau richtig ist.
Wenn die richtige Zahl eingegeben wurde, Schleife abbrechen und die Anzahl der benötigten Durchgänge ausgeben.
"""
# Guessing Game: First, generate a random number between 1 and 20.
# In a loop, prompt the user for input and then indicate whether the input is too small, too large, or exactly right.
# If the correct number is entered, break the loop and display the number of attempts required.


```

```python
import random

count = 0
zufallszahl = random.randint(1,20)
print(zufallszahl)
while True:
    n = int(input("guess a nummer (1-20):"))
    count += 1
    if n < zufallszahl:
        print(f"input nummer {n} ist klein ")
    elif n > zufallszahl:
        print(f"input nummer {n} ist groß ")
    else:
        print("input nummer", n, "ist genau richtig")
        print(f"it took {count} attempts")
        break

```

```python
"""""
Aufgabe 2:
Die Lösung der Aufgabe 1 als Funktion definieren. Dann 5 Runden des Spieles durch Aufrufen der Funktion durchführen und insgesamt eine Liste mit den Returnwerten aufbauen.
Z.B. Liste [4, 2, 5, 3, 2] bedeutet, man hat in den fünf Spielrunden jeweils 4, 2, 5, 3 und 2 Durchgänge benötigt.
"""""
# Define the solution to Task 1 as a function. Then play 5 rounds of the game by calling the function and build a list of the return values.
# For example, the list [4, 2, 5, 3, 2] means that it took 4, 2, 5, 3, and 2 attempts, respectively, in each of the five rounds of the game.

```

```python
import random
def guessing_game():
    count = 0
    zufallszahl = random.randint(1,20)
    print(zufallszahl)

    while True:
        n = int(input("guess a nummer:"))
        count += 1
        if n < zufallszahl:
            print(f"input nummer {n} ist klein ")
        elif n > zufallszahl:
            print(f"input nummer {n} ist groß")
        else:
            print("input nummer", n, "ist genau richtig")
            return count

guessing_game()

result = []

for i in range(5):
    print(f"\nRound {i+1}")      #'\n is a newline character. It tells Python:Start a new line here”
    result = guessing_game()
    result.append(result)

print("\nAll results:", result)

# interne Info über funktion wird ausgegeben, meist uninteressant
print(guessing_game())



```
