```python
# The unterschied: Methode gegenüber globaler Funktion
"""
what is the difference between Method and global function

Good question—this is a core concept in Python and programming in general. Let’s make it very clear.

🔹 Simple difference
✅ Global function

A function that exists on its own, not attached to any object.

def greet(name):
    return "Hello " + name

print(greet("Alice"))

👉 greet is a global function

✅ Method

A function that belongs to an object (or class) and is called using .

text = "hello"
print(text.upper())

👉 upper() is a method of a string object

🔹 Key difference in one sentence
Function → works independently
Method → works on an object
🔹 Side-by-side example
Function:
len([1, 2, 3])

👉 len() is a global function

Method:
"hello".upper()

👉 upper() belongs to the string "hello"

🔹 How they are called
Type	How you call it
Function	function(x)
Method	object.method()
🔹 Important idea (very helpful)

A method always knows the object it belongs to.

Example:

text = "hello"
text.upper()

Behind the scenes, Python is roughly doing:

str.upper(text)

So the object (text) is passed automatically.

🔹 Real examples you already saw

From your earlier question:

file = Path("example.txt")
file.exists()
exists() → method (belongs to Path object)
print(file)
print() → global function
🔹 Analogy

Think of it like this:

Function = a tool in your toolbox 🔧
Method = a button on a specific machine 🎛️

You:

use a tool anywhere → function
press a button on a machine → method
🔹 Quick rule to remember

👉 If you see a dot (.), it’s a method

obj.something()  # method

👉 If not, it’s usually a function

something(obj)   # function

If you want, I can show you how to create your own methods inside a class—that’s the next step and makes this concept really click.
"""
```
